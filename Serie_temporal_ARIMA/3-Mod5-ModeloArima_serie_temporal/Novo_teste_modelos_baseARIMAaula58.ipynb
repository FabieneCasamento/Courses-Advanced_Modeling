# Faremos outro tipo de transforma√ß√£o: Faremos a raiz c√∫bica, a mais indicada, que pode com dados zero ou negativos


```python
serie3 = (serie)**(1/3)
serie3
```

```python
import scipy.stats as stats
stats.probplot(serie3, dist="norm", plot=plt)
plt.title("Normal QQ plot")
plt.show()

e, p =stats.shapiro(serie3)
print('Estat√≠stica de teste: {}'.format(e))
print('p-valor: {}'.format(p))


if p > 0.05:
   print(f'Ho = distribui√ß√£o normal p-valor: {p} > 0,05')
elif p <=0.05:
   print(f'Ha = distribui√ß√£o n√£o normal p-valor: {p} <= 0,05')
   if p < 0.000001:
       print(f'\p-valor: {p} => muito abaixo de 0,05')

```

### **Estacionaridade**






```python
import statsmodels.tsa.stattools



kpss = statsmodels.tsa.stattools.kpss(serie3)
print('Estat√≠tica do teste: {:.4f}'.format(kpss[0]))
print('p_valor: {:.4f}'.format(kpss[1]))
print('Valores Cr√≠ticos:')
for chave, valor in kpss[3].items():
   print('{}: {:.4f}'.format(chave, valor))
    
valor_critico = kpss[3]['5%']
if kpss[0] > kpss[3]['5%']:
    print(f'Ha = n√£o √© estacion√°rio: estat√≠stica do teste {kpss[0]} > valor cr√≠tico {valor_critico}')
elif kpss[0] < kpss[3]['5%']:
    print(f'Ho = √© estacion√°rio:  estat√≠stica do teste {kpss[0]}  < valor cr√≠tico {valor_critico}')
print()
```

# autocorrela√ß√£o


## plot_acf: fun√ß√£o autocorrela√ß√£o

verifica a autocorrela√ß√£o entre os valores vizinhos

## plot_pacf: fun√ß√£o de autocorrela√ß√£o parcial

autocorrela√ß√£o entre os valores intercalados 

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf



# no exemplo a distribui√ß√£o n√£o √© normal, mas haver√° outros testes

plot_acf(serie3, lags=20)
plt.show()


plot_pacf(serie3, lags=20)
plt.show()
plot_pacf(serie, lags=20)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# separado e depois juntar, n√£o significa que seja a um modelo melhor.

modelo_arima = ARIMA(serie3, order = (9,0,8))

resultado_arima = modelo_arima.fit()
print(resultado_arima.summary())
```

### **An√°lise dos Res√≠duos**

```python
residuos_arima = resultado_arima.resid
```

```python

import matplotlib.dates as mdates


# Certificando-se de que o √≠ndice √© uma s√©rie temporal (caso n√£o esteja), formatando a data
if not pd.api.types.is_datetime64_any_dtype(residuos_arima.index):
    residuos_arima.index = pd.to_datetime(residuos_arima.index, format='%Y-%m-%d')  # Converte para ano-m√™s



# Criando a figura e o gr√°fico
plt.figure(figsize=(10, 5))

# Plotando a s√©rie

# s√©rie do res√≠duo
plt.plot(residuos_arima.index, residuos_arima.values, label='res√≠duo da s√©rie 3 ARIMA (9,0,8)', color='blue')


# Adicionando um t√≠tulo ao gr√°fico
plt.title("Res√≠duo da serie 3 ARIMA (9,0,8)")

# Definindo r√≥tulos dos eixos
plt.xlabel("Dados mensais")
plt.ylabel("Res√≠duo")

# Rotacionando os r√≥tulos das datas no eixo x para melhorar a visualiza√ß√£o
plt.xticks(rotation=75)


# Formatando o eixo x para mostrar a cada 18 dias
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=18))  # Colocar um marcador a cada 18 dias
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%Y'))  # Formatando para exibir ano e m√™s

plt.axhline(0, color='green', linestyle='--', linewidth=0.8)  # Linha horizontal em y=0
plt.axhline(residuos_arima.mean(), color='red', linestyle='--', linewidth=0.8)  # Linha horizontal em y=0


# Adicionando a legenda
plt.legend()

# Exibindo o gr√°fico
plt.tight_layout()
plt.show()
```

## comparar de forma de tabela


```python
# Arredondar as colunas de interesse
resultado_merge_ARIMA['diferen√ßa dado previsto - real(mm)'] = resultado_merge_ARIMA['diferen√ßa dado previsto - real(mm)'].round(2)
resultado_merge_ARIMA['anomalia de chuva real(mm)'] = resultado_merge_ARIMA['anomalia de chuva real(mm)'].round(2)
resultado_merge_ARIMA['anomalia de chuva prevista(mm)'] = resultado_merge_ARIMA['anomalia de chuva prevista(mm)'].round(2)

# Criar a coluna 'Resultados' com setas e texto
resultado_merge_ARIMA['Resultados'] = [
    'Previs√£o Subestimada (‚¨áÔ∏è)' if comparacao < 0 
    else 'Previs√£o Superestimada (‚¨ÜÔ∏è)' if comparacao > 0 
    else 'Previs√£o Precisa'
    for comparacao in resultado_merge_ARIMA['diferen√ßa dado previsto - real(mm)']
]

# Fun√ß√£o para aplicar cor ao texto na coluna 'Resultados'
def color_results(val):
    if 'Subestimada' in val:
        return 'color: red'
    elif 'Superestimada' in val:
        return 'color: blue'
    return 'color: green'

# Fun√ß√£o para aplicar cor de fundo: vermelho para valores < 0, azul para valores > 0
def background_results(val):
    if val < 0:
        return 'background-color: crimson'
    elif val > 0:
        return 'background-color: mediumblue'
    return 'background-color: green'  # Nenhuma cor para valores iguais a zero
    
#https://www.delftstack.com/pt/howto/python/colors-in-python/

# Aplicar estilos
resultado_merge_ARIMA_styled = (
    resultado_merge_ARIMA.style
    .applymap(color_results, subset=['Resultados'])  # Aplicar cor ao texto na coluna 'Resultados'
    .applymap(background_results, subset=['anomalia de chuva real(mm)', 'anomalia de chuva prevista(mm)'])  # Cor de fundo para colunas especificadas
)

# Formatar colunas num√©ricas com duas casas decimais
lista_col = list(resultado_merge_ARIMA.columns)
coluna = lista_col[1:7]
resultado_merge_ARIMA_styled = resultado_merge_ARIMA_styled.format({col: "{:.2f}" for col in coluna})

# Exibir o resultado estilizado
resultado_merge_ARIMA_styled
# o modelo previu bem a anomalia de chuva para o segundo, terceiro e o quinto m√™s. 
```

# Faremos outro tipo de transformação: Faremos a raiz cúbica, a mais indicada, que pode com dados zero ou negativos


```python
serie3 = (serie)**(1/3)
serie3
```

```python
import scipy.stats as stats
stats.probplot(serie3, dist="norm", plot=plt)
plt.title("Normal QQ plot")
plt.show()

e, p =stats.shapiro(serie3)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))


if p > 0.05:
   print(f'Ho = distribuição normal p-valor: {p} > 0,05')
elif p <=0.05:
   print(f'Ha = distribuição não normal p-valor: {p} <= 0,05')
   if p < 0.000001:
       print(f'\p-valor: {p} => muito abaixo de 0,05')

```

### **Estacionaridade**






```python
import statsmodels.tsa.stattools



kpss = statsmodels.tsa.stattools.kpss(serie3)
print('Estatítica do teste: {:.4f}'.format(kpss[0]))
print('p_valor: {:.4f}'.format(kpss[1]))
print('Valores Críticos:')
for chave, valor in kpss[3].items():
   print('{}: {:.4f}'.format(chave, valor))
    
valor_critico = kpss[3]['5%']
if kpss[0] > kpss[3]['5%']:
    print(f'Ha = não é estacionário: estatística do teste {kpss[0]} > valor crítico {valor_critico}')
elif kpss[0] < kpss[3]['5%']:
    print(f'Ho = é estacionário:  estatística do teste {kpss[0]}  < valor crítico {valor_critico}')
print()
```

# autocorrelação


## plot_acf: função autocorrelação

verifica a autocorrelação entre os valores vizinhos

## plot_pacf: função de autocorrelação parcial

autocorrelação entre os valores intercalados 

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf



# no exemplo a distribuição não é normal, mas haverá outros testes

plot_acf(serie3, lags=20)
plt.show()


plot_pacf(serie3, lags=20)
plt.show()
plot_pacf(serie, lags=20)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# separado e depois juntar, não significa que seja a um modelo melhor.

modelo_arima = ARIMA(serie3, order = (9,0,8))

resultado_arima = modelo_arima.fit()
print(resultado_arima.summary())
```

## comparar de forma de tabela


```python
# Arredondar as colunas de interesse
resultado_merge_ARIMA['diferença dado previsto - real(mm)'] = resultado_merge_ARIMA['diferença dado previsto - real(mm)'].round(2)
resultado_merge_ARIMA['anomalia de chuva real(mm)'] = resultado_merge_ARIMA['anomalia de chuva real(mm)'].round(2)
resultado_merge_ARIMA['anomalia de chuva prevista(mm)'] = resultado_merge_ARIMA['anomalia de chuva prevista(mm)'].round(2)

# Criar a coluna 'Resultados' com setas e texto
resultado_merge_ARIMA['Resultados'] = [
    'Previsão Subestimada (⬇️)' if comparacao < 0 
    else 'Previsão Superestimada (⬆️)' if comparacao > 0 
    else 'Previsão Precisa'
    for comparacao in resultado_merge_ARIMA['diferença dado previsto - real(mm)']
]

# Função para aplicar cor ao texto na coluna 'Resultados'
def color_results(val):
    if 'Subestimada' in val:
        return 'color: red'
    elif 'Superestimada' in val:
        return 'color: blue'
    return 'color: green'

# Função para aplicar cor de fundo: vermelho para valores < 0, azul para valores > 0
def background_results(val):
    if val < 0:
        return 'background-color: crimson'
    elif val > 0:
        return 'background-color: mediumblue'
    return 'background-color: green'  # Nenhuma cor para valores iguais a zero
    
#https://www.delftstack.com/pt/howto/python/colors-in-python/

# Aplicar estilos
resultado_merge_ARIMA_styled = (
    resultado_merge_ARIMA.style
    .applymap(color_results, subset=['Resultados'])  # Aplicar cor ao texto na coluna 'Resultados'
    .applymap(background_results, subset=['anomalia de chuva real(mm)', 'anomalia de chuva prevista(mm)'])  # Cor de fundo para colunas especificadas
)

# Formatar colunas numéricas com duas casas decimais
lista_col = list(resultado_merge_ARIMA.columns)
coluna = lista_col[1:7]
resultado_merge_ARIMA_styled = resultado_merge_ARIMA_styled.format({col: "{:.2f}" for col in coluna})

# Exibir o resultado estilizado
resultado_merge_ARIMA_styled
# o modelo previu bem a anomalia de chuva para o segundo, terceiro e o quinto mês. 
```

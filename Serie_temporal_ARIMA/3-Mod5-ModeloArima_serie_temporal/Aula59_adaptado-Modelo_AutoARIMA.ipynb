# Faremos outro tipo de transforma√ß√£o: Faremos a raiz c√∫bica, a mais indicada, que pode com dados zero ou negativos


```python
serie3 = (serie)**(1/3)
serie3
```

```python
import scipy.stats as stats
stats.probplot(serie3, dist="norm", plot=plt)
plt.title("Normal QQ plot")
plt.show()

e, p =stats.shapiro(serie3)
print('Estat√≠stica de teste: {}'.format(e))
print('p-valor: {}'.format(p))


if p > 0.05:
   print(f'Ho = distribui√ß√£o normal p-valor: {p} > 0,05')
elif p <=0.05:
   print(f'Ha = distribui√ß√£o n√£o normal p-valor: {p} <= 0,05')
   if p < 0.000001:
       print(f'\p-valor: {p} => muito abaixo de 0,05')

```

### **Estacionaridade**






```python
import statsmodels.tsa.stattools



kpss = statsmodels.tsa.stattools.kpss(serie3)
print('Estat√≠tica do teste: {:.4f}'.format(kpss[0]))
print('p_valor: {:.4f}'.format(kpss[1]))
print('Valores Cr√≠ticos:')
for chave, valor in kpss[3].items():
   print('{}: {:.4f}'.format(chave, valor))
    
valor_critico = kpss[3]['5%']
if kpss[0] > kpss[3]['5%']:
    print(f'Ha = n√£o √© estacion√°rio: estat√≠stica do teste {kpss[0]} > valor cr√≠tico {valor_critico}')
elif kpss[0] < kpss[3]['5%']:
    print(f'Ho = √© estacion√°rio:  estat√≠stica do teste {kpss[0]}  < valor cr√≠tico {valor_critico}')
print()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# separado e depois juntar, n√£o significa que seja a um modelo melhor.

modelo_arima = ARIMA(serie3, order = (9,0,8))

resultado_arima = modelo_arima.fit()
print(resultado_arima.summary())
```

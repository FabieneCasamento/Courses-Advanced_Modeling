```python
# Importação das bibliotecas tradicionais
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Criação da Série

```python
# Criação da série tempora e testando logo depois a sazonalidade. 
# dados aleatórios 
# rodar semente aleatória (.seed), mas para que várias vezes rodadas, deêm os mesmos valores. 
np.random.seed(10)



# gerar pelo np. random e uma distribuição, para mostrar uma média =0, 
#um desvio padrão = 1, e 41 valores gerados, um exemplo para 41 anos. 

dados1 = np.random.normal(0,1,41)
dados1
```

```python
# Gerando os ados para criar a série, salvando como dataframe. 
dados1 = pd.DataFrame(dados1)
dados1.head()
```

```python
# renomear a coluna
dados1.columns = ['valores']
dados1.head()
```

```python
print('(linha, coluna)')
dados1.shape
```

```python
# preparar a série temporal, gerando os anos como os anos.
#período, quantidade de anos, freq = Year
indice = pd.date_range('1980', periods = len(dados1), freq = 'Y')
indice
```

```python

#após coar ps dadps como dataframe e os anos como index, geramos uma série temporal.
serie1 = pd.Series(dados1['valores'].values, index = indice)
serie1
```

```python
# Plotando a série
plt.plot(serie1)

# Adicionando um título ao gráfico
plt.title("Meu Gráfico de Série 1")
          
# Plotando a série com uma legenda
plt.plot(serie1, label="Série 1")

# Adicionando um título ao gráfico
plt.title("Meu Gráfico de Série 1")

plt.xlabel("anos")
plt.ylabel("Série normal")

plt.legend()

plt.show()
```

```python
# Plotando a série com uma legenda
print('os dados não tem muita variância e uma média constante, aparentemente. Porém parece que tem uma pequena dispersão')
plt.title("Meu Gráfico de Série 1")

plt.plot(serie1, label="Série 1")


plt.axhline(y=serie1.mean(), color='red', linestyle='--', label="Linha Horizontal da média da série")

plt.xlabel("anos")
plt.ylabel("Série normal")

# Adicionando a legenda
plt.legend()

# Exibindo o gráfico
plt.show()
```

```python
# Uma série que tem uma tendência crescente ou decrescente, claramente não é estacionária. 
```

## **TESTES DE ESTACIONARIDADE**

## Iremos fazer dois tipos de testes de estacionaridade. 

alguns usam o p-valor, e a estatítica do teste. 

```python
import statsmodels.tsa.stattools
```

Teste KPSS (Kwiatkowski-Phillips-Schmidt-Shin)

Ha rejeitar a hipótese nula

Ha = não é estacionária: estatística do teste > valor crítico.

H0 aceitar a hipótese nula

Ho = é estacionária:  estatística do teste <= valor crítico.

```python
# geração de métricas, pelo dicionário. 
statsmodels.tsa.stattools.kpss(serie1)
```

```python
kpss = statsmodels.tsa.stattools.kpss(serie1)
print('Estatítica do teste: {:.4f}'.format(kpss[0]))
print('p_valor: {:.4f}'.format(kpss[1]))
print('número de lags: {}'.format(kpss[2]))
print('Valores Críticos:')
for chave, valor in kpss[3].items():
   print('{}: {:.4f}'.format(chave, valor))
```

# nessa análise, em exceção , não usamos os dados do lag. 

Estatítica do teste: 0.0686

p_valor: 0.1000

número de lags: 0

Valores Críticos:

10%: 0.3470  

5%: 0.4630  geralmente usamos o nível de confiança de 5% e com o valor crítico de 0,4630.

2.5%: 0.5740

1%: 0.7390

Comparar o valor estatítico do teste com o valor crítico referente a ível de confiança de 5%





Estatítica do teste: 0.0686 ficou menor que 5%: 0.4630 

Estatítica do teste: 0.0686  < 5%: 0.4630 

H0 aceitar a hipótese nula

Ho = é estacionária:  estatística do teste <= valor crítico.
 o que representa que a nossa série é estacionária. 

# outro teste é o Teste df (Dickey Fuller)

### Teste df (Dickey Fuller): Que é um processo inverso do KPSS (Kwiatkowski-Phillips-Schmidt-Shin)



Teste df (Dickey Fuller)

Hipótese nula (h0) NÃO é estacionária


Ho = não é estacionária: estatística do teste > valor crítico.

Hipótese alternativa (Ha) é estacionária


Ha = é estacionária:  estatística do teste <= valor crítico.



```python
df = statsmodels.tsa.stattools.adfuller(serie1)
print('Estatística do teste: {:.4f}'.format(df[0]))
print('p_valor: {:.8f}'.format(df[1]))
print('Valores Críticos:')
for chave, valor in df[4].items():
   print('{}: {:.4f}'.format(chave, valor))
```

Estatística do teste: -5.8642

p_valor: 0.00000034

Valores Críticos:

1%: -3.6056

5%: -2.9371 # por padrão usaremos o nível de 5%, com o valor crítico 

10%: -2.6070



Estatística do teste: -5.8642 < 5%: -2.9371 # por padrão usaremos o nível de 5%, com o valor crítico 

Ha = é estacionária:  estatística do teste <= valor crítico.


Este teste afirma que os dados são estacionarios, isto é, rejeitamos a hipótese nula e aceitamos a hipótese alternativa.

## **TESTES DE ESTACIONARIDADE**

```python
import statsmodels.tsa.stattools
```

Teste KPSS (Kwiatkowski-Phillips-Schmidt-Shin)

Ha = não é estacionária:  **estatística do teste > valor crítico**.

Ho = é estacionária: **estatística do teste <= valor crítico**


```python
kpss2 = statsmodels.tsa.stattools.kpss(serie2)
print('Estatítica do teste: {:.4f}'.format(kpss2[0]))
print('p_valor: {:.4f}'.format(kpss2[1]))
print('Valores Críticos:')
for chave, valor in kpss2[3].items():
   print('{}: {:.4f}'.format(chave, valor))
```


Estatítica do teste: 1.6513  > valor crítico: 5%: 0.4630


Então, realmente o valor não é estacionário.

Ha = não é estacionária:  **estatística do teste > valor crítico**.

Teste df (Dickey Fuller)

Ho = não é estacionária: estatística do teste > valor crítico.

Ha = é estacionária:  estatística do teste <= valor crítico.

```python
df2 = statsmodels.tsa.stattools.adfuller(serie2)
print('Estatítica do teste: {:.4f}'.format(df2[0]))
print('p_valor: {:.8f}'.format(df2[1]))
print('Valores Críticos:')
for chave, valor in df2[4].items():
   print('{}: {:.4f}'.format(chave, valor))
```


Estatítica do teste: 0.8154   >    5%: -2.8840


Ho = não é estacionária: estatística do teste > valor crítico.


Mostramos as características e as análises da série, para depois começar a fazer um correção. Então há um processo chamado de <b>diferenciação</b> que aplicamos quando a série <b>não é estacionária</b>

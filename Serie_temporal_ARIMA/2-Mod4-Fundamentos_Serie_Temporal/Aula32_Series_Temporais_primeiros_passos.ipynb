# Para verificar se os dados são uma distribuição normal.

```python
# biblioteca matemática, com a parte estatística stats

import scipy.stats as stats
```

```python

# gráfico Quantil Quantil 
# gráfico de probabilidades, que é um método gráfico para comparar duas distribuições de probabilidade
# https://pt.wikipedia.org/wiki/Gr%C3%A1fico_Q-Q
stats.probplot(serie1, dist="norm", plot=plt)
plt.title("Normal QQ plot")
plt.show()
```


x= quantiles teoricos, e o y = nossos valores - no caso normalizados e ordenados.

# Linha de referência em vermelho, e os pontos em azul, tem que coincidir com a linha vermelha. Se isso acontecer, quer dizer que nós temos uma distribuição normal, como se pode ver na figura acima. 



Por outro lado, quanto mais disperso estiver esses pontos em azul, com relação à linha vermelha pior seria, pior com relação à análise de anormalidades.

pontos azuis próximos da linha vermelha, indica uma distribuição normal. => na verificação gráfica

# Agora na verificação numérica

CRITÉRIOS:

NÍVEL DE SIGNIFICÂNCIA DE 0,05 ou 5% (MAIS UTILIZADO)

A distribuição NÃO vai ser normal quando o p foi menor ou igual a 0,05 | p <= 0,05. 

estatística do teste

```python
e, p = stats.shapiro(serie1)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

p- valor ficou maior que 0,05, o que deixa a distribuição como normal

**Teste Shapiro-Wilk**

CRITÉRIOS:

NÍVEL DE SIGNIFICÂNCIA DE 0,05 ou 5% (MAIS UTILIZADO)

QUANDO p > 0,05 (distribuição normal).

```python
e, p = stats.shapiro(serie3)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
#p > 0,05 (distribuição normal).
```

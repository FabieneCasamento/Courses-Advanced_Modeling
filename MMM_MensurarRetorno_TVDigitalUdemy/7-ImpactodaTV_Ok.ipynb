```python
import pandas as pd
df_Base_fusca = pd.read_excel("05_MMM_Base-Fusca.xlsx", sheet_name= "Base Fusca", skiprows=4, decimal=',', usecols=  ['Month', 'Facebook', 'Programática',
       'Youtube', 'UOL', 'Outros', 'Facebook.1', 'Programática.1', 'Youtube.1',
       'UOL.1', 'Outros.1'])
df_Base_fusca.columns= [ 'Month', 'Facebook_investimento', 'Programática_investimento', 'Youtube_investimento', 'UOL_investimento', 'Outros_investimento', 'Facebook_leads', 'Programática_leads', 'Youtube_leads',
       'UOL_leads', 'Outros_leads']


df_Base_fusca['Offline_investimento'] = 0
# ação offline da tv
df_Base_fusca.loc[df_Base_fusca['Month'] == '2016-01-01','Offline_investimento'] = 1
df_Base_fusca.loc[df_Base_fusca['Month'] == '2016-02-01','Offline_investimento'] = 1
df_Base_fusca.loc[df_Base_fusca['Month'] == '2017-05-02','Offline_investimento'] = 1

df_Base_fusca['total_leads']= df_Base_fusca['Facebook_leads'] + df_Base_fusca['Programática_leads'] + df_Base_fusca['Youtube_leads']+ df_Base_fusca['UOL_leads'] +df_Base_fusca['Outros_leads']

df_Base_fusca.head()
```

# estátistico

```python
# não significa que o R2 baixo teremos um modelo que não fique bem ajustado. Vamos analisar mais a fundo.

import statsmodels.formula.api as smf
# dados completos soma_geral_treino
df_estatistico = df_Base_fusca[:16]
# Modelo com fórmula
# Modelo com fórmula corrigida (usando + para somar variáveis)
modelo = smf.ols("total_leads ~ Facebook_investimento + Programática_investimento + Youtube_investimento + UOL_investimento+ Outros_investimento + Offline_investimento", data=df_Base_fusca).fit()

# Estatísticas de regressão
resumo = modelo.summary()

resumo
```

# Analisar o P>|t|	-> olhar o valor

```python
import statsmodels.api as sm

# ANOVA
anova = sm.stats.anova_lm(modelo, typ=2)

print('\nANOVA:')
display(anova)
```

```python
# Para calcular o erro padrão geral (RMSE ou erro padrão residual) entre o eixo X (Investimentos) e Y (Leads)
# Erro padrão
import numpy as np
from sklearn.metrics import mean_squared_error
# Selecionar as variáveis independentes (X) e dependente (y)
x = df_estatistico[[
    'Facebook_investimento',
    'Programática_investimento',
    'Youtube_investimento',
    'UOL_investimento',
    'Outros_investimento',
    'Offline_investimento'
]]
y = df_estatistico['total_leads']    
# Predição
y_pred2 = modelo.predict(x)    
# Erro padrão geral (RMSE)
rmse = np.sqrt(mean_squared_error(y, y_pred2))
print(f'Erro padrão geral -  Erro Quadrático Médio (RMSE): {rmse:.2f}')
```

# olhar a significância dependendo do contexto. 

```python
# Coeficientes (incluindo o intercepto)
coeficientes = modelo.params

# Intercepto (constante)
coef_intercepto = coeficientes['Intercept']

# Coeficientes das variáveis independentes
coef_facebook = coeficientes['Facebook_investimento']
coef_programatica = coeficientes['Programática_investimento']
coef_youtube = coeficientes['Youtube_investimento']
coef_uol = coeficientes['UOL_investimento']
coef_outros = coeficientes['Outros_investimento']
coef_offline = coeficientes['Offline_investimento']

print('dados estatísticos do investimento e leads do Fusca')
print("Intercepto (Coeficientes Interseção) :", coef_intercepto)
print("Facebook:", coef_facebook)
print("Programática:", coef_programatica)
print("YouTube:", coef_youtube)
print("UOL_investimento:",coef_uol)
print("Outros:", coef_outros)
print("Offline:", coef_offline)
```

```python
# R-quadrado
r2 = modelo.rsquared

# R-quadrado ajustado
r2_ajustado = modelo.rsquared_adj

# Exibir
print(f"R²: {r2:.4f} ")
print(f"R² ajustado: {r2_ajustado:.4f}")
```

✅ Qualidade do Modelo
R² = 0.533

O modelo explica 53,3% da variação no número total de leads com base nos investimentos.

Interpretação: Há uma relação moderada entre os investimentos e os leads. Porém, ainda há 46,7% da variação não explicada.

R² ajustado = 0.279

Corrige o R² para o número de variáveis no modelo.

Interpretação: A qualidade do modelo piora ao ajustar para o número de variáveis (indicando possível overfitting ou variáveis pouco relevantes).

=========================================================================================================================
- <b>Overfitting (ou sobreajuste)</b> é quando um modelo estatístico ou de machine learning se <b>ajusta bem demais aos dados de treinamento</b>, capturando não apenas os padrões reais, mas também o <b>ruído e as variações aleatórias</b>. Isso faz com que o modelo funcione muito bem nos dados que já viu, mas tenha um <b>desempenho ruim em dados novos</b>.

📉 Sinais de Overfitting:
R² muito alto nos dados de treino, mas baixo nos dados de teste.

Muitos parâmetros ou variáveis irrelevantes no modelo.

Erro muito pequeno no treino, mas grande no teste.

🧠 Analogia: Overfitting é como decorar todas as respostas de um simulado em vez de aprender o conteúdo — você vai bem na prova simulada, mas mal na prova real.




F-statistic = 2.095 | p = 0.136
Teste para saber se o modelo como um todo é estatisticamente significativo.

### Interpretação: Não é estatisticamente significativo ao nível de 5% (p > 0.05). Há chance de que os resultados sejam por acaso.

# Ideia Lasso 

📌 Objetivo

Reduzir overfitting e selecionar automaticamente as variáveis mais relevantes entre os tipos de investimento.

✅ 1. Preparar os dados
Certifique-se de que:

```python
df_Base_fusca_lasso= df_Base_fusca[:18]
df_Base_fusca_lasso
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np

# Suponha que df seja seu DataFrame com as variáveis
X = df_Base_fusca_lasso[['Facebook_investimento', 'Programática_investimento', 'Youtube_investimento',
        'UOL_investimento', 'Outros_investimento', 'Offline_investimento']]
y = df_Base_fusca_lasso['total_leads']

# Separar treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train, y_test
```

✅ 2. Treinar modelo com Lasso

```python
# Lasso com penalização (alpha controla a regularização)
modelo_lasso = Lasso(alpha=0.1)
modelo_lasso.fit(X_train, y_train)
```

✅ 3. Avaliar desempenho

```python
# Previsão
y_pred = modelo_lasso.predict(X_test)

# Métricas
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"R²: {r2:.4f}")
print(f"RMSE: {rmse:.2f}")
```

✅ 4. Interpretar coeficientes

```python
# Coeficientes de cada variável
coeficientes = pd.Series(modelo_lasso.coef_, index=X.columns)
print(coeficientes)

# Ver quais variáveis foram eliminadas (coef = 0)
variaveis_eliminadas = coeficientes[coeficientes == 0].index.tolist()
print("Variáveis eliminadas pelo Lasso:", variaveis_eliminadas)
```

🧠 Dica de ajuste fino

Use LassoCV para encontrar o melhor valor de alpha automaticamente:




```python
from sklearn.linear_model import LassoCV

modelo_cv = LassoCV(cv=5).fit(X_train, y_train)
print(f"Melhor alpha: {modelo_cv.alpha_}")
```

### Marketing Mix Modeling (MMM)

Equação matemática com o objetivo utilizar como variáveis as ações publicitárias, com base em investimento ou não (como eventos). 

MOdelo de regressão, com facilidade de manusear os modelos

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Eixo X e Y
x = soma_geral_treino['Soma Investimentos'].values.reshape(-1, 1)
y = soma_geral_treino['Soma Leads'].values

# Regressão linear
modelo = LinearRegression()
modelo.fit(x, y)

# Predição
y_pred = modelo.predict(x)
r2 = modelo.score(x, y)  # R-quadrado

# Plot
plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='blue', label='Dados reais')
plt.plot(x, y_pred, color='red', linewidth=2, label='Regressão Linear')

# Texto com R²
plt.text(x.min(), y.max()*0.95, f'$R^2$ = {r2:.3f}', fontsize=12, color='black')

plt.xlabel('Soma Investimento (R$)')
plt.ylabel('Soma Leads')
plt.title('Regressão Linear entre Investimento e Leads (gráfico de dispersão)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Coeficientes
print(f'Coeficiente angular (slope): {modelo.coef_[0]:.4f}  (Coeficientes investimento)')
print(f'Intercepto: {modelo.intercept_:.4f}  (Coeficientes Interseção)')


# valores previstos = (coeficientes interseção) + (coeficientes investimento)*investimentos

Coeficientes_intercepto = modelo.intercept_
Coeficientes_investimento = modelo.coef_[0]
```

```python
import statsmodels.formula.api as smf
# dados completos soma_geral_treino
# Modelo com fórmula
modelo = smf.ols("Leads ~ Investimentos", data=soma_geral_treino).fit()

# Estatísticas de regressão
resumo = modelo.summary()

resumo
```

O valor-p (ou p-value) indica a probabilidade de que os resultados observados tenham ocorrido por acaso, assumindo que a hipótese nula (de que não há relação entre as variáveis) seja verdadeira.

✅ Interpretação do valor-p
Valor-p	Interpretação prática

"≤ 0,05" :	Resultado estatisticamente significativo. Rejeita-se a hipótese nula. A variável explicativa (ex: investimento) tem influência sobre a variável resposta (ex: Leads).

"> 0,05": Resultado não significativo. Não se rejeita a hipótese nula. Pode não haver relação clara entre as variáveis.


Um valor-p pequeno para um coeficiente (ex: do investimento) significa que o coeficiente é relevante e não é zero por acaso.

O valor ideal é menor que 0,05, principalmente em ciências sociais e negócios.

Em resumo: quanto menor o valor-p, maior a evidência de que a variável tem impacto real.

```python
import statsmodels.api as sm

# ANOVA
anova = sm.stats.anova_lm(modelo, typ=2)

print('\nANOVA:')
print(anova)
```

```python
#  colocar a parte treino jan/16 e fev/17.
df_base_porType_Car_treino = df_base_porType_Car[:14]
df_base_porType_Car_treino
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Eixo X e Y
x = df_base_porType_Car_treino['Investimentos'].values.reshape(-1, 1)  # investimento 
y = df_base_porType_Car_treino['Leads'].values # leads

# Regressão linear
modelo = LinearRegression()
modelo.fit(x, y)

# Predição
y_pred = modelo.predict(x)
r2 = modelo.score(x, y)  # R-quadrado

# Plot
plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='blue', label='Dados reais')
plt.plot(x, y_pred, color='red', linewidth=2, label='Regressão Linear')

# Texto com R²
plt.text(x.min(), y.max()*0.95, f'$R^2$ = {r2:.3f}', fontsize=12, color='black')

plt.xlabel('Investimento (R$)')
plt.ylabel('Leads')
plt.title(f'Regressão Linear entre Investimento e Leads (gráfico de dispersão) {busca_tipo_car}')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Coeficientes
print(f'Coeficiente angular (slope) em {busca_tipo_car}: {modelo.coef_[0]:.4f}  (Coeficientes investimento)')
print(f'Intercepto: {modelo.intercept_:.4f}  (Coeficientes Interseção) em {busca_tipo_car}')


# valores previstos = (coeficientes interseção) + (coeficientes investimento)*investimentos

Coeficientes_intercepto = modelo.intercept_
Coeficientes_investimento = modelo.coef_[0]



```

```python
# não significa que o R2 baixo teremos um modelo que não fique bem ajustado. Vamos analisar mais a fundo.

import statsmodels.formula.api as smf
# dados completos soma_geral_treino
# Modelo com fórmula
modelo = smf.ols("Leads ~ Investimentos", data=df_base_porType_Car_treino).fit()

# Estatísticas de regressão
resumo = modelo.summary()

resumo
```

O valor-p (ou p-value) indica a probabilidade de que os resultados observados tenham ocorrido por acaso, assumindo que a hipótese nula (de que não há relação entre as variáveis) seja verdadeira.

✅ Interpretação do valor-p
Valor-p	Interpretação prática

"≤ 0,05" :	Resultado estatisticamente significativo. Rejeita-se a hipótese nula. A variável explicativa (ex: investimento) tem influência sobre a variável resposta (ex: Leads).

"> 0,05": Resultado não significativo. Não se rejeita a hipótese nula. Pode não haver relação clara entre as variáveis.


Um valor-p pequeno para um coeficiente (ex: do investimento) significa que o coeficiente é relevante e não é zero por acaso.

O valor ideal é menor que 0,05, principalmente em ciências sociais e negócios.

Em resumo: quanto menor o valor-p, maior a evidência de que a variável tem impacto real.

📊 Resumo da Regressão Linear (OLS) - Resultado do Fusca

Objetivo: Avaliar a relação entre Investimentos e Leads.

1. Coeficientes:

Intercepto (1319.70)

Interpretação: Se o investimento fosse zero, o modelo estima que haveria cerca de 1.320 leads.

Obs: P-valor = 0.082 → não é estatisticamente significativo (acima de 0.05).

Coeficiente de Investimentos (0.0022)

Interpretação: A cada R$ 1 adicional investido, espera-se um aumento de 0.0022 leads, ou seja, 2,2 leads a cada mil reais investidos.

Obs: P-valor = 0.061 → próximo de significativo, mas não passa no critério padrão de 5%.


--------------------------------------------------------------

📈 Estatísticas do Modelo v

R-quadrado = 0.262

Interpretação: O modelo explica 26,2% da variação no número de leads — não é alto, mas indica alguma associação.

R-quadrado ajustado = 0.201

Leva em conta o número de variáveis. Confirma que o modelo tem explicação modesta.

F-estatística = 4.266, P = 0.061

Testa se o modelo como um todo é significativo. Quase significativo, mas está ligeiramente acima do limite convencional (0.05).

--------------------------------------------------------------

🧠 Conclusão Resumida

Há uma tendência de que o investimento impacte os leads, mas os resultados não são estatisticamente significativos ao nível de 5%.

O modelo explica apenas uma parte da variação — pode haver outras variáveis importantes ausentes.

É um modelo inicial promissor, mas precisa ser refinado ou complementado com mais dados ou variáveis.

# Entendendo a sazonalidade 

A sazonalidade  uma caractersstica especfica das sries temporais que as diferencia de outros tipos de dados. 

A sazonalidade  basicamente a representao de algum ritmo ou comportamento, seja da natureza ou da atividade humana, que se expressa atravs de ciclos. 

Ento ns representamos em sries temporais esses ciclones em formas de sazonalidade. 

Nesse captulo, ns vamos entender como que o Prophet interpreta a sazonalidade por padro. 

Como odemos ajust-las ou mesmo adicionar novas sazonalidades. 

Um primeiro aspecto que ns temos que entender sobre sazonalidade seria o seu tipo.

Ns temos a sazonalidade aditiva e a sazonalidade multiplicativa. 

+ Na aditiva no decorrer da srie no altera a sua amplitude, ou seja tem uma amplitude fixa com relao ao dado. 

+ J na multiplicativa, h umaumento ou uma mudana de amplitude ao decorrer da srie. 

# faremos testes com os dados, mudando a sazonalidades. 

```python
import pandas as pd
df_air_passengers = pd.read_csv('../Modulo4/AirPassengers.csv')
display(df_air_passengers)
```

```python
df_air_passengers['Month'] = pd.to_datetime(df_air_passengers['Month'])
df_air_passengers.columns =['ds','y']
df_air_passengers
```

# Aditivos

```python
#importar a biblioteca
from prophet import Prophet
#instanciado o modelo
model_a = Prophet(seasonality_mode ='additive') # Por padro os dados j so aditivos
model_a.fit(df_air_passengers)
future = model_a.make_future_dataframe(periods = 12 * 5, freq= 'MS')
forecast = model_a.predict(future)
# Visualizar o Forecast
import matplotlib.pyplot as plt
fig = model_a.plot(forecast)
plt.show() 
```

### o prophet no conseguiu acompanhar o aumento da amplitude dos dados. No final, a previso ficou com uma amplitude menor do que o dado deveria ter. E isso permance para o perodo projetado  frente. Ento esse  um exemplo claro de uma srrie qu tem que ser modelada com a sazonalidade no modo multiplicativo. Por que a amplitude dos dados vai se alterando no decorrer da srie. 

# multiplicativos 

```python
#importar a biblioteca
from prophet import Prophet
#instnciado o modelo
model_m = Prophet(seasonality_mode ='multiplicative')
model_m.fit(df_air_passengers)
future = model_m.make_future_dataframe(periods = 12 * 5, freq= 'MS')
forecast = model_m.predict(future)
# Visualizar o Forecast
import matplotlib.pyplot as plt
fig = model_m.plot(forecast)
plt.show() 
```

quando olhamos a srie modelada com a sazonaidade multiplicativa, ns vemos que o prophet ele conseguiu se ajustar ao crescimento da variao dos dados ao longo da srie. Ns vemos uma previso mais coerente algo que se estende at para o perodo que no temos dados de treinamento. 

diferena dos multiplicativos e aditivos. 

# componentes aditivos 

```python
fig2 = model_a.plot_components(forecast)
fig2.show()
```

# componentes multiplicativos

```python
fig2 = model_m.plot_components(forecast)
fig2.show()
```

Os grficos acima de sazonalidade e tendncia esto parecidos, mas quando trabalhamos em uma sazonalidade aditiva, os valores so tratados de forma absoluta aqui na sazonalidade. Na parte anula, tem alguns valores fixos que so acrescentados  nossa tendncia de forma fixa. Isse reflete em um aspecto que pode gerar problema em algumas sries temporais, que , no caso dela, prev valores negativos, quando isso no seria possvel. 

Imagine se essa tenncia aqui fosse uma tendncia de queda, como a nossa sazonalidade estaria com um valor fixo, o que aconteceria. Conforme a tendncia fosse caindo, conforme a tendncia se aproximasse de zero, ele adicionaria esse valor absoluto  tenncia e possivelmente a gente teria valores negativos na tendncia. Sendo que isso NO seria possvel observando no mundo real. 


Quando olhamos para a sazonalidade multiplicativa, ns vemos que h um valor relativo, ou seja, a sazonalidade sempre vai ser relativa ao valor que temos presente na tendncia. Ento se a tendncia est aumentando, a amplitude da nossa sazonalidade vai aumentar de forma relativa  nossa tendncia, de forma que se ns tivssemos uma tendncia de queda, comentado no cenrio anterior, de possveis valores negativos, isso j no aconteceria na srie multiplicativa. 

Por que, conforme a nossa tendncia fosse se aproximando de zero, a amplitude tambm iria reduzindo de forma que ela nunca efetivamente ficaria negativa.

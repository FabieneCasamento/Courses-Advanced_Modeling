# Vamos entender como trabalhar com dados subdiários, ou seja, com uma granularidade menor que um dia, assim como os dados mensais. 

Assim como os dados mensais, nós também precisaremos fazer um ajuste para que o Prophet entenda corretamente esses dados. 

Vamos usar uma base diferente da análise anterior estudada.

# divvy_hourly.csv é uma base de dados de uma empresa que aluga biciletas. São dados de hora a hora, com cada corrida. 

```python
import pandas as pd

df_divvy_bikes = pd.read_csv('divvy_hourly.csv')
display(df_divvy_bikes)
```

```python

# Vamos fazer as transformações 

df_divvy_bikes['date']=pd.to_datetime(df_divvy_bikes['date'])
df_divvy_bikes.columns = ['ds','y']
df_divvy_bikes
```

```python
from prophet import Prophet
model = Prophet(seasonality_mode = 'multiplicative')
model.fit(df_divvy_bikes)
```

```python

# dados por hora, freq='h', com a granulidade de hora a hora são 365*24, 1 hora com 24 horas
future = model.make_future_dataframe(periods = 365*24 , freq='h')
```

```python
forecast = model.predict(future)
```

```python
forecast.head()
```

```python

import matplotlib.pyplot as plt

fig = model.plot(forecast)

plt.show()
```


Analisando, o primeiro ponto é que o Prophet não conseguiu se adequar totalmente a amplitude desses dados. Talvez esses dados sejam  parte outliers, mas há uma incidência muito grande. O mais provável é que ele não conseguiu se adequar aqui a esse senóide que os dados fazem. 

Ou seja, os dados reais são os pontos escuros e o previsto o azul. O intervalo de confiança ficou muito grande, indicando que o modelo, neste caso, não tem uma segurança com relação à previsão desses dados.

Para ele garantir 95% de chance do resultado previsto está dentro do intervalo de confiança, na atual figura é mostrado que ele precisa determinar um intervalo de confiança muito grande, que dessa forma seria efetivamente muito difícil ele errar e fazer uma previsão fora desse intervalo. Então isso nos dá indícios de que o Prophet não conseguiu capturar a complexidade total dos nossos dados, de forma com que ele conseguisse projetar de forma confiante. 

# Para entender melhor podemos plotar os componentes internos. 

```python
fig2 = model.plot_components(forecast)
fig2.show()
```

## Capturou a tendência e 3 sazonalidades: semanal, anual e diário. É a primeira vez que estamos trabalhando com dados com granularidade menores que um dia. 


Ela só vai aparecer nesse cenário. 

A sazonalidade só vai aparecer quando a gente tiver dois (2) períodos completos daquelas granularidade específica. Por exemplo, a diária só aparece quando a gente tiver dois (2) DIAS de informações completas, de informação sudiária. A anual é da mesma forma, só vai aparecer se a gente tiver 2 anos completos. Para que ela consgia captura a diferença ano contra ano e assim entender alguma sazonalidade.  

Analisando de forma mais detalhada a tendência. Neste exemplo ainda dá para notar que há um ruído de sazonalidade dentro da tendência, ou seja, o prophet não conseguiu extrair por padrão toda a sazonalidade da tendência. 



Então esse ruído aqui faz com que o Prophet se sinta inseguro em prever a tendência. Nós vimos aqui um intervalo de confiança abrindo bastante nesta previsão. 


### Nós vamos precisar de parâmetros adicionais no prophet para que ele consiga capturar sazonalidades de forma adequada e ter uma tendência mais linear. 



O modelo conseguiu capturar uma sazonalidade semanal muito bem. Tendo um padrão muito bem definido, onde há baixas nos finais de semanas e picos nos horários de semana. A sazonalidade anual, mas ainda existem ruídos na tendência anual (verficamos que ainda há ondas presentes no gráfico yearly), aind aprecisa capturar a tendência anual de forma mais adequada. 
Sazonalidade diária está bem estabelecida, onde consegue entender que há baixas nos períodos noturno e picos nos horáris que vão trabalhar e outro pico na volta para casa após o trabalho, no final da tarde e no início da noite.

Neste caso, entendemos como recepcionar esses dados hora a hora, da forma mais adequada no Prophet. 

Isso é feito através da especificação do parâmetro frequência e da correta determinação dos períodos. 

### Dados com Gaps Regulares

Lidando com lacunas nos dados

Imagine que você tem uma base diária, mas ela só possui dados nos dias úteis, ou seja, nos final de semana haverá um gap regular onde nenhum final de semana terá dados. Esse é um exemplo de gap regular, que ele é constante. 
Um exemplo de gap irregular seria se você historicamente tivesse dados no final de semana e um ou outro dia você não tivesse dados. Isso seria no caso uma 'missing data'. 

Isso é uma abordagem diferente que a gente pode lidar com esses dados, para preencher eses dados. Para que não atrapalhe nossa série temporal. No caso, <b> o Prophet já consegue lidar automaticamente com essas 'missing datas esporádicos'</b>. 



### Mas, quando há um gap regular, a gente precisa trabalhar de uma forma diferente. Outro cenário de gap regular seria: imagine uma base subdiaria, onde você tem os dados hora a hora, mas você só tem dados do horário de expediente, que é justamente o que a gente vai tentar simular com a nossa base da divvy bikes, que a gente trabalhou no nosso último exemplo. 

Então vou filtrar a base da divvy bikes para que ela só tenha dados no horário de expediente. 

```python
# filtragem para dados no horário comercial
df = df_divvy_bikes[(df_divvy_bikes['ds'].dt.hour >= 8) & (df_divvy_bikes['ds'].dt.hour <18)]
df.head(25)
```

```python
# passando para o df modificado
model = Prophet(seasonality_mode = 'multiplicative')
model.fit(df)
model
```

```python
# dados por hora, freq='h', com a granulidade de hora a hora são 365*24, 1 hora com 24 horas
future = model.make_future_dataframe(periods = 365*24 , freq='h')
```

```python
forecast = model.predict(future)
forecast.head()
```

```python

fig = model.plot(forecast)

plt.show()
```

Com isso, conseguimos ver que a previsão ficou muito incoerente com rlação aos dados que nós temos na nossa base histórica. 


Para ajustar isso, é um processo semelhante a qual a gente utilizou nos exercícios anteriores. Em que a gente determinava uma frequência, mas ela é um pouco mais personalizada. o Prophet não tem os acessos fora do horário de expediente, mas no nosso dataframe futuro, nós estamos gerando uma base que tem todo esse range de horários. Então a abordamengem é eliminar desse dataframe os horários para o qual o prophet não tem dados de treinamentos para fazer a previsão corretamente. 

# Como iremos fazer, é apresentar o mesmo filtro que nós fizemos no DataFrame da divvy bikes, que será filtrado no make_future_dataframe

```python
#dados filtrados entre as 8 e às 18 horas.
future2 = future[(future['ds'].dt.hour >= 8) & (future['ds'].dt.hour <18)]
forecast = model.predict(future2)
forecast.head()
```

```python
fig = model.plot(forecast)

plt.show()
```


estamos vendo agora que a previsão está muito mais coerente. Em via de regra, independente da granularidade ou do gap que houver, o prophet vai conseguir fazer a previsão.

Mas, <b>você tem que garantir que o dataframe futuro contenha períodos que o Prophet tem dados históricos para fazer uma previsão</b>. Por exemplo, se sua base não tivesse finais de semana, vocẽ teria que filtrar aqui do dataframe futuro as datas de finais de semana. 
Isso faria com que o prophet produzisse um resultado coerente.

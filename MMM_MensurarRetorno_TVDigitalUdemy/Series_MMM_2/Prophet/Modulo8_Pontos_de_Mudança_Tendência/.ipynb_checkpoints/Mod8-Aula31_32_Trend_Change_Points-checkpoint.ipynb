
As séries temporais no mundo real possuem diversas mudanças em sua trajetória, de forma que puramente linear. Não é possível traçar uma reta que represente toda a natureza das séries temporais. Você deve ter notado que, em alguns exemplos, a tendência não era uma reta perfeita. Isso porque o Porphet tem a capacidade de receber mudanças no comportamento da tendência e traçar uma nova tendência a partir daquele ponto. Essas mudanças de tendências no Prophet são chamadas de change points. 


Entender como funciona o comportamento padrão do Prophet referente a essas mudanças e como podemos ajustá-las manualmente, quando entendemos que não está se comportando de forma mais adequada. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from prophet.plot import add_changepoints_to_plot
df = pd.read_csv('../Modulo5/divvy_daily.csv')
display(df)
```

```python
df = df[['date', 'rides']]
df['date'] = pd.to_datetime(df['date'])
df.columns = ['ds', 'y']
display(df)
```

#### Função auxiliar que vai nos ajudar a visualizar onde que os Change Points estão localizados: 

```python

import numpy as np

def set_changepoints(df, n_changepoints, changepoint_range):

    df = df.sort_values('ds').reset_index(drop=True)

    hist_size = int(np.floor(df.shape[0] * changepoint_range))
    if n_changepoints + 1 > hist_size:
        n_changepoints = hist_size - 1
        print('n_changepoints greater than number of observations. Using {}.'.format(n_changepoints))
    if n_changepoints > 0:
        cp_indexes = np.linspace(0,
                                 hist_size - 1,
                                 n_changepoints + 1).round().astype(int)
        changepoints = df.iloc[cp_indexes]['ds'].tail(-1)
    else:
        # set empty changepoints
        changepoints = pd.Series(pd.to_datetime([]), name='ds')
    return changepoints
```



Vamos visualizar como que o Prophet distribui o Change points em uma série temporal por padrão


Por padrão o Prophet colocará 25 pontos de mudanças potenciais espaçadas uniformemente nos primeiros 80% dos dados. 

Antes mesmo de determinar se aquele é efetivamente um ponto de mudança, repare, é um ponto de mudança potencial. Os pontos potenciais estão indicados por linhas tracejadas na vertical. 

```python
changepoints = set_changepoints(df, 25, .8)
#Por padrão o Prophet colocará 25 pontos de mudanças potenciais espaçadas uniformemente nos primeiros 80% dos dados.
changepoints
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(df['ds'], df['y'], c='#0072B2')
for cp in set_changepoints(df,25, .8):
    plt.axvline(x=cp, c='gray', ls='--')
plt.xlabel('Date')
plt.ylabel('Rides per day')
plt.show()
```

Vamos instanciar o nosso modelo com uma configuração padrão. É nessa etapa que o Prophet determinará a magnitude de mudanças de cada um desses potenciais Change Points, e irá determinar se eles efetivamente representam uma alteração repentina na inclinação da nossa tendência. 

```python
model = Prophet(seasonality_mode='multiplicative',yearly_seasonality=4)
model.fit(df)
forecast = model.predict()
forecast
```


Vamos plotar os change points oficiais.


```python

fig = model.plot(forecast)
add_changepoints_to_plot(fig.gca(), model, forecast)
plt.show()
```

### As linhas tracejadas verticais são os change points que o Prophet entendeu como significativo. 

A figura descreve que entre uma linha e outra tracejada houve uma mudança significativa na tendência. Por mais que visualmente ela não seja tão evidente matematicamente, a diferença que nós temos de inclinação entre essa reta e essa se tornou significativamente para o Porphet. 
A magnitude de cada um daqueles 25 pontos potenciais de mudança é armazenada no model params dentro do item <b>delta</b>

```python

print(model.params['delta'])
```

## Repare que esses valores foram normalizados de mode que seus valores absolutos não têm um efeito sentidos para nós, mas suas magnitudes relativas sim. 

Como esses números são calculados por meio de um processo de otimização em vez de uma equação determinística. Você pode chegar a valores exatos diferentes. Mas os expoentes sempre deve ser aproximadamente os mesmos. A maioria dessas maginitudes tem um expoente de -8 com -9, o que significa que na notação padrão você deve deslocar o ponto decimal desse número 8 ou 9 dígitos para esquerda, o que significa que esses números estão muito próximos de 0. 
Você pode visualizar essas magnitudes através desse código abaixo. 

Na imagem abaixo temos alguns elementos. 

```python

ylims = fig.gca().get_ylim()

fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'k'
add_changepoints_to_plot(ax1, model, forecast, trend=True)
ax1.set_xlabel('ds')
ax1.set_ylabel('y', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(ylims[0] + 5000, ylims[1] - 7000)
ax1.grid(False)

ax2 = ax1.twinx()

color = '#0072B2'
ax2.set_ylabel('Trend rate change')
ax2.bar(changepoints, model.params['delta'][0], width=30, color=color, alpha=.75, label='Trend rate change')
ax2.tick_params(axis='y', labelcolor='k')
ax2.axhline(y=0, c=color, ls='--')
ax2.axhline(y=.01, c='gray', ls=':')
ax2.axhline(y=-.01, c='gray', ls=':')
ax2.set_ylim(top=-ax2.get_ylim()[0], bottom=ax2.get_ylim()[0])

ax2.legend()

fig.tight_layout()
ax1.set_zorder(1)
ax1.patch.set_visible(False)
plt.show()
```



A linha sólida é a tendência; as linhas tracejadas são os chang points significativos; as barras  são os valores das magnitudes do ponto de mudança, ou seja, quanto que a inclinação de um segmento desses 25 cortes; as barras verticais são as magnitudes dos pontos de mudanças, ou seja, o quanto que aquela tendência se inclinou de um segmento para outro. 

Novamente, a maior parte dessas magnitudes é quase 0, portanto, não aparece no gráfico onde a linha tracejada horizontal azula indica um ponto 0 dessa magnitude. 



As barras que se posicionam para cima são pontos de mudança com magnitude positiva, ou seja, uma inclinação positiva da nossa tendência. E as barras que se estendem para baixo significam que a nossa tendência teve um declínio. 

Em resumo, essa é a forma que o Prophet identifica mudanças de tendência e a forma como ele trabalha por padrão. 

Futuramente, vamos entender como ajustar os change points conforme a necessidade. 

## Regularização de Change Points

Vamos aprender com alterar o comportamento automático do Prophet com relação ao Change Points.

O Prophet distribui 25 change points potenciais nos primeiros 80% dos dados. Depois separa quais deles tem um valor significativo de alterações com relação à marcação do Change Points anterior.

A título de curiosidade, podemos alterar o número de change points através do argumento n_changepoints, que pode ser usado na instanciação do modelo. Vamos carregar o nosso modelo com o parâmetro Change Points = 5.

```python
model = Prophet(seasonality_mode='multiplicative', yearly_seasonality=4, n_changepoints=5)
```

Visualizar como que seria a distribuição desses 5 Change Points; Setar o parâmetro como 5, na função Change Points;

# Essa seria a distribuição simulada dos change points que o Prophet faz com essa configuração.  com 0.8 ou 80% dos dados

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['ds'], df['y'], c='#0072B2')
for cp in set_changepoints(df, 5, .8):
    plt.axvline(x=cp, c='gray', ls='--')
plt.xlabel('Date')
plt.ylabel('Rides per day')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['ds'], df['y'], c='#0072B2')
for cp in set_changepoints(df, 5, .5):
    plt.axvline(x=cp, c='gray', ls='--')
plt.xlabel('Date')
plt.ylabel('Rides per day')
plt.show()
```


Uma outra configuração possível é também alterar o range no qual o Prophet está aplicando os Change Points, ou os potenciais Change Points. No caso, através desse parâmetro a função, conseguimos especificar que, por exemplo, se apliqe os possíveis change poins apenas na primeira metade dos dados. (0.5).

set_changepoints(df, 5, .5):


Em ambos os casos, é importante lembrar que você não está instruindo o Prophet onde colocar exatamente os pontos de mudança, são apenas pontos de mudança potenciais. Onde se o Prophet achar que é significativo a alteração de tendência entre um ponto e outro, ele vai eleger aquela marcação como um change point oficialmente. Além disso, é imporante lembrar que o Prophet nunca estabelecerá um ponto de mudança no Futuro. É por isso que, por padrão, o Prophet apenas utiliza os 80 primeiros porcento dos dados, para evitar que ele escolha um ponto de mudança ruim, muito próximo do término da série temporal, onde que ele tenha poucos pontos de dados futuros para eventualmente corrigir aquele erro de cálculo referente a tendência.
No geral, as configurações padrões do Prophet funcionará bem na maioria dos casos. Se você achar que o Prophet está se ajustando demais ou insuficientemente aos pontos de mudança, é melhor controlar isso através da regularização. 


Nesse ponto, temos uma informação interessante, por padrão, tanto o seasonality_prior_scale, quanto o holidays_space_scale  holidays_prior_scale foram definidos como 10. 

```python

model = Prophet(seasonality_mode='multiplicative',yearly_seasonality=4,changepoint_prior_scale=1)
model.fit(df)
forecast = model.predict()
fig = model.plot(forecast)
add_changepoints_to_plot(fig.gca(), model, forecast)
plt.show()
```

Porém, o changepoint_prior_scale, no entanto, é definido como por padrão 0,05. Mas, assim como acontece na sazonalidade nos feriados, aumentar esse vlaor tornará a tendência mais flexível e diminuí-lo tornará menos flexível. Os valores razoáveis normalmente estão na faixa entre .001 e .05 

Neste caso, vamos instanciar nvamente nosso modelo, só que alterando nosso prior_scale para 1, ou seja, aumentando ele. 


Isso permite que o Prophet tenha uma maior flexibilidade, ou seja, estará muito mais sensível ao que ele considera uma alteração na tendência e, por isso, identifica mais pontos de mudança como significativo. Aqui, como podemos ver, dos 25 que estão determinados como padrão nesse modelo, ele identificou vários como pontos de mudança real de tendência. 


model = Prophet(seasonality_mode='multiplicative',yearly_seasonality=4,changepoint_prior_scale=.007)
model.fit(df)
forecast = model.predict()
fig = model.plot(forecast)
add_changepoints_to_plot(fig.gca(), model, forecast)
plt.show()


Por outro lado, vamos ver o que acontece quando regularizados com muita rigidez. Vamos reduzir para .007, executando novamente. E com isso vemos que o modelo se tornou muito menos sensível a alteração de tendência. Elegendo apenas 4 Change Points, como Change Points reais. 



# Especificando Change Points Personalizados
# Especificand locais de pontos de mudanças personalizados


Iremos entender como sugerir para o Prophet um Change Point de forma manual, em uma data específica, que você analista entende que causou um impacto na sua série temporal 





Neste exemplo, vamos usar uma base que contém o número de likes do instagram de um jogador de futebol, chamado James Rodrigues. Carregando em um dataframe. 

## *Especificando locais de pontos de mudança personalizados**"""


```python

df = pd.read_csv('instagram_jamesrodriguez10.csv')
display(df)
df['Date'] = pd.to_datetime(df['Date'])
df.columns = ['ds', 'y']
df
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x=df['ds'], y=df['y'], c='#0072B2')
plt.ticklabel_format(style='plain', axis='y')
plt.xlabel('Date')
plt.ylabel('Likes per photo')
plt.show()
```

No segundo bloco temos o código responsável pela visualização dessa série temporal. Analisando essa série, vemos que há momentos que ela possui movimentos abruptos. Especificamente aqui, em 2014, onde houveram Copas do Mundo e ele foi um jogador de destaque. E um terceiro pico aqui, mais sutil, quando ele foi tranferido para o Bayern de Munique. 

Uma primeira coisa que vamos fzer para modelar essa série vai ser colocar o pico que vemos como feriado, para que o prophet entenda que ali é um movimento especial.

Assim criamos ab aixo um dataframe para cada movimento abrupto. 
Copa de 2014, copa de 2018 e a mudança para o Bayern de Munique.

```python
wc_2014 = pd.DataFrame({'holiday': 'World Cup 2014','ds':pd.to_datetime(['2014-06-12']),'lower_window': 0,'upper_window': 31})
wc_2018 = pd.DataFrame({'holiday': 'World Cup 2018','ds':pd.to_datetime(['2018-06-14']),'lower_window': 0,'upper_window': 31})
signing = pd.DataFrame({'holiday': 'Bayern Munich','ds':pd.to_datetime(['2017-07-11']),'lower_window': 0,'upper_window': 14})
#unido todos os dataframes
special_events = pd.concat([wc_2014, wc_2018, signing])
special_events
```

```python
# isso para depois passarmos como feriados para o nosso modelo
```

Neste bloco abaixo nós temos uma lista de datas que são as datas nas quais a gente deseja especificar um change points para o Prophet. Elas estão imediatamente antes e depois desses movimentos abruptos, e a nossa intenção com isso é de que, chegando nesse endpoint, o Prophet calcule uma nova reta para que nesse intervalo entre Change Points o Prophet recalcule uma tendência específica para aquele segmento, para aquele movimento anormal, mas que no resto da série, onde temos um movimento não anormal, que ele tenha uma tendência mais limpa, sem ruído desse pico abrupto. 

assim, vamos executar a lista.

```python

changepoints = ['2014-06-12',
'2014-07-13',
'2017-07-11',
'2017-07-31',
'2018-06-14',
'2018-07-15']
```

Assim, especificamos o changepoints nos parâmetros do modelo. 

```python
model = Prophet(seasonality_mode='multiplicative',
  holidays=special_events,
  yearly_seasonality=False,
  weekly_seasonality=False,
  changepoint_prior_scale=1,
  changepoints=changepoints)

# executando a instanciação
```

```python
model.fit(df)
forecast = model.predict()
forecast.head()
```

```python
fig = model.plot(forecast)
add_changepoints_to_plot(fig.gca(), model, forecast)
plt.show()
fig2 = model.plot_components(forecast)
plt.show()
```

Na primeira figura, perceba que nós adicionamos antes e depois daquele pico, os Change Points, nesse momento, ele recalcula uma tendência que é muito forte, de alta, mas depois ele tem uma liberdade para calcular uma tendência com os dados que estão dentro da normalidade. E é dessa forma que nós conseguimos, de alguma forma, isolar esses momentos mais abruptos que existem nas séries temporais que nós vemos por aí. 

E é dessa forma nós sugerimos os Change Poins Manuais para o Prophet.


Vamos aprender um hack para poder acelear o processo de validação cruzada da sua série temporal, que por vezes pode demorar um pouco. 

Quando executamos uma validação cruzada, muitas iterações podem acontecer, até que você obtenha um resultado final. 

O prophet utiliza a opção de paralelização Processo por padrão e faz com que o prophet paralise a classe Processes poll executor que faz com que o Python rode cada iteração em processos independentes. E essa opção atende bem. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from prophet.plot import add_changepoints_to_plot

df= pd.read_csv('online_retail.csv')
display(df)
df['date'] = pd.to_datetime(df['date'])
df.columns = ['ds', 'y']
df
```

```python
model = Prophet(yearly_seasonality=4)
model.fit(df)
```

```python
future = model.make_future_dataframe(periods=365) # previsão para mais 365 dias
future
```

```python
forecast = model.predict(future)
forecast
```

```python
# para fazer a validação cruzada, iremos importar a cross_validation

from prophet.diagnostics import cross_validation
```

df_cv = cross_validation(model,
                        horizon='90 days',
                        period='30 days',
                        initial='730 days', parallel = 'processes')


```python
df_cv
```


Opções parâmetro 'parallel': processes; dask, threads, None

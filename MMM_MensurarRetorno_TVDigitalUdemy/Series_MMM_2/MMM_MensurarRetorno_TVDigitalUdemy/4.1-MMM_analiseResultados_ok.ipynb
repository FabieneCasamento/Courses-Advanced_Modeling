```python
import pandas as pd
df_Media_Plan = pd.read_excel("01_MMM.xlsx", sheet_name= "Media Plan", skiprows=3, decimal=',', usecols=['Unnamed: 2'	,'JULHO',	'AGOSTO'])
df_Media_Plan
```

```python
# resultados leads
df_resultado = pd.read_excel("01_MMM.xlsx", sheet_name= "Resultados",  skiprows=3, usecols= ['Month' ,'Fusca', 'Fiat 147',
       'Passat Hatch', 'Passat Sedan', 'Tempra', 'Buggy', 'Kombi', 'Rural'])
df_resultado
```

```python
# formação base
# Revertendo o pivot com melt
df_melted_investimento = pd.melt(df_investimento, id_vars=['Month'], var_name='Carros_Nameplate', value_name='Investimentos')
display(df_melted_investimento.head(10))
```

```python
# Revertendo o pivot com melt
df_melted_resultado= pd.melt(df_resultado, id_vars=['Month'], var_name='Carros_Nameplate', value_name='Leads')
display(df_melted_resultado.head(10))

```

```python
df_base_formatada = df_melted_investimento.merge(df_melted_resultado, on=['Month','Carros_Nameplate'])
df_base_formatada['ROI(%)']= (df_base_formatada['Leads']/df_base_formatada['Investimentos']) *100
df_base_formatada.head(40)
```

```python
# Somar as tabelas digitais de julho e agosto.

df_Media_Plan_digital = df_Media_Plan[16:25]
df_Media_Plan_digital
```

```python
soma_digital_plano_midia_julho2017 = df_Media_Plan_digital['JULHO'].sum()
soma_digital_plano_midia_agosto2017 = df_Media_Plan_digital['AGOSTO'].sum()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
```

```python

baseline = []
for busca_tipo_car in list(df_base_formatada['Carros_Nameplate'].unique()):
    # coleta dos dados
    print(busca_tipo_car)
    df_base_porType_Car  = df_base_formatada.loc[df_base_formatada['Carros_Nameplate']==busca_tipo_car].reset_index(drop=True)
    df_base_porType_Car['Investimentos formatada'] = df_base_porType_Car['Investimentos'].apply(lambda x: f'{x:,.2f}')
    display(df_base_porType_Car)
    #-----------------------------------------------------------
    # Busca do Share digital do tipo de Car de Interesse
    df_share_digital= df_Media_Plan[27:].reset_index(drop=True).rename(columns={'Unnamed: 2':'Share (DIGITAL)'})
    display(df_share_digital)
    df_Share_digitalporType_Car =  df_share_digital.loc[df_share_digital['Share (DIGITAL)']==busca_tipo_car].reset_index(drop=True)
    display(df_Share_digitalporType_Car)
    # ----------------------------------------------------------
    # plano da média
    print('Ainda se soma todos os dados do Digital, mas será considerado apenas a porcentagem de interesse do tipo de carro')
    display(df_Media_Plan_digital)
    # ----------------------------------------------------------
    #---- cálculo da média dos planos e investimentos de interesse e previsão e previsão
    #df_Share_digitalporType_Car.info()
    porcentagem_share_Jul = df_Share_digitalporType_Car['JULHO'].values[0]
    porcentagem_share_Ago = df_Share_digitalporType_Car['AGOSTO'].values[0]
    print(f"""
    
    Investimentos no digital para {busca_tipo_car}:
    
    multiplicando o investimento de Julho {soma_digital_plano_midia_julho2017:,.2f} e multiplicando com o share de {porcentagem_share_Jul*100}% do mesmo mês
    Soma Plano Mídia de 07/2017 de investimentos R$ {soma_digital_plano_midia_julho2017*porcentagem_share_Jul:,.2f}
    
    multiplicando o investimento de Agosto {soma_digital_plano_midia_agosto2017:,.2f} e multiplicando com o share de {porcentagem_share_Ago*100}% do mesmo mês
    Soma Plano Mídia de 08/2017 de investimentos R$ {soma_digital_plano_midia_agosto2017*porcentagem_share_Ago:,.2f}
       
    """)
    # ------------------------------------------------------------------
    # separação para o treino
    #  colocar a parte treino jan/16 e fev/17.
    df_base_porType_Car_treino = df_base_porType_Car[:14]
    display(df_base_porType_Car_treino)


    # -------------------------------------------------------------------
    # -------- plotagem busca da regressão-------------------------------------------------
    print('plotagem regressão')
    # Eixo X e Y
    x = df_base_porType_Car_treino['Investimentos'].values.reshape(-1, 1)  # investimento 
    y = df_base_porType_Car_treino['Leads'].values # leads
    
    # Regressão linear
    modelo = LinearRegression()
    modelo.fit(x, y)
    
    # Predição
    y_pred = modelo.predict(x)
    r2 = modelo.score(x, y)  # R-quadrado
    
    # Plot
    plt.figure(figsize=(10, 5))
    plt.scatter(x, y, color='blue', label='Dados reais')
    plt.plot(x, y_pred, color='red', linewidth=2, label='Regressão Linear')
    
    # Texto com R²
    plt.text(x.min(), y.max()*0.95, f'$R^2$ = {r2:.3f}', fontsize=12, color='black')
    
    plt.xlabel('Investimento (R$)')
    plt.ylabel('Leads')
    plt.title(f'Regressão Linear entre Investimento e Leads (gráfico de dispersão) {busca_tipo_car}')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    # Coeficientes
    print(f'Coeficiente angular (slope) em {busca_tipo_car}: {modelo.coef_[0]:.4f}  (Coeficientes investimento)')
    print(f'Intercepto: {modelo.intercept_:.4f}  (Coeficientes Interseção) em {busca_tipo_car}')
    
    
    # valores previstos = (coeficientes interseção) + (coeficientes investimento)*investimentos
    
    Coeficientes_intercepto = modelo.intercept_
    Coeficientes_investimento = modelo.coef_[0]
    # ----------------------------------------------------------------------------
    # não significa que o R2 baixo teremos um modelo que não fique bem ajustado. Vamos analisar mais a fundo.

    import statsmodels.formula.api as smf
    # dados completos soma_geral_treino
    # Modelo com fórmula
    modelo = smf.ols("Leads ~ Investimentos", data=df_base_porType_Car_treino).fit()
    
    # Estatísticas de regressão
    resumo = modelo.summary()
    print('resumo estatístico')
    display(resumo)
    
    # ANOVA
    anova = sm.stats.anova_lm(modelo, typ=2)
    print('resumo estatístico 2')
    print('\nANOVA:')
    print(anova)
   # print('*'*280)
    # ----------------------------------------
    # Para calcular o erro padrão geral (RMSE ou erro padrão residual) entre o eixo X (Investimentos) e Y (Leads)
    #from sklearn.metrics import mean_squared_error
    # X e Y
    x = df_base_porType_Car_treino['Investimentos']
    y = df_base_porType_Car_treino['Leads']
    
    # Predição
    y_pred2 = modelo.predict(x)
    
    # Erro padrão geral (RMSE)
    rmse = np.sqrt(mean_squared_error(y, y_pred2))
    print(f'Erro padrão geral (RMSE): {rmse:.2f}')
    #Esse valor te dá o erro médio em unidades da variável dependente (Leads) — quanto menor, melhor o ajuste do modelo.


    # ----------------------------------------------------------------------------
    # cálulo dos valores previstos 
    # ----------------------------------------------------------------------------
    soma_digital_plano_midia_julho2017_type_car = soma_digital_plano_midia_julho2017*porcentagem_share_Jul
    soma_digital_plano_midia_agosto2017_type_car = soma_digital_plano_midia_agosto2017*porcentagem_share_Ago
        
    Valores_Previstos_julho2017_type_car = Coeficientes_intercepto + Coeficientes_investimento* soma_digital_plano_midia_julho2017_type_car
    Valores_Previstos_agosto2017_type_car = Coeficientes_intercepto + Coeficientes_investimento* soma_digital_plano_midia_agosto2017_type_car
       
    print(f"""
    Valores previstos por Mês do {busca_tipo_car}
    Valores previstos de leads de 07/2017 em {busca_tipo_car}: {Valores_Previstos_julho2017_type_car:,.2f}
    Valores previstos de leads de 08/2017  em {busca_tipo_car}: {Valores_Previstos_agosto2017_type_car:,.2f}
    """)
    # ----------------------------------------------------------------------------------
    # ajustar todos os dados 
    df_base_porType_Car['Valores_Previstos'] = Coeficientes_intercepto + Coeficientes_investimento* df_base_porType_Car['Investimentos'] 
    display(df_base_porType_Car.head(7))
    print('objetivo, é fazer uma previsão para o mês de março e abril. Na parte do teste')
    df_Type_car_teste = df_base_porType_Car[14:16].reset_index(drop=True)
    display(df_Type_car_teste) # março e abril de 2017 para o teste.
    # ----------------------------------------------------------------------
    # Erro = 1 - valor previstos/leads
    print('Erro = 1 - valor previstos/leads')
    leads_TESTE = df_Type_car_teste['Leads'].sum() 
    valores_previstos_TESTE = df_Type_car_teste['Valores_Previstos'].sum()
    Erro = 1 - (valores_previstos_TESTE/leads_TESTE)
    print(f'Erro em {busca_tipo_car}: {Erro*100:.1f}%')  

    # comparação para o teste março e abril
    # ----------------------------------------------------
    leads_TESTE_marco_abril = leads_TESTE
    valores_previstos_TESTE_leads_marco_abril = valores_previstos_TESTE

    # ----------------------------------------------------
    # plotagem comparacao
    display()

    plt.figure(figsize=(12, 6))
    
    # Plotar os leads reais
    plt.plot(df_base_porType_Car['Month'], df_base_porType_Car['Leads'], marker='o', label='Leads Reais', color='blue')
    
    # Plotar os valores previstos
    plt.plot(df_base_porType_Car['Month'], df_base_porType_Car['Valores_Previstos'], marker='x', linestyle='--', label='Valores Previstos', color='red')
    
    plt.title(f'Leads Reais vs Valores Previstos da base de treinos + teste {busca_tipo_car}')
    plt.xlabel('Mês')
    plt.ylabel('Quantidade de Leads')
    plt.xticks(rotation=45)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()


    # registros para colocar na tabela 
    # Erro - variação, e RMSE erro padrão geral (RMSE ou erro padrão residual)
    registros = [busca_tipo_car, r2,Erro,rmse, leads_TESTE_marco_abril, valores_previstos_TESTE_leads_marco_abril ]
    baseline.append(registros)

    print('*'*270)



```

# variação: abs(1 - previsto/real)

```python
colunas = ['carro', 'R2', 'erro(variação)', 'RMSE', 'leads reais', 'valores previstos leads']
df_baseline = pd.DataFrame(baseline, columns=colunas)
df_baseline['erro(variação) %'] = df_baseline['erro(variação)']*100
#variação = abs(1 - previsto/real)
df_baseline['variação %'] =  abs(1 - df_baseline['valores previstos leads']/ df_baseline['leads reais'])*100
df_baseline
```

# cáculo do share - relembrando

```python

soma_geral = df_base_formatada.groupby('Month').agg({'Leads': 'sum', 'Investimentos':'sum', 'ROI(%)': 'mean' }).reset_index()
soma_geral = soma_geral.rename(columns={'Leads': 'Soma Leads', 'Investimentos': 'Soma Investimentos', 'ROI(%)':'Media ROI(%)'})

soma_geral['Soma Investimentos formatada'] = soma_geral['Soma Investimentos'].apply(lambda x: f'{x:,.2f}')
display(soma_geral)


print(f"Soma total dos leads {soma_geral['Soma Leads'].sum():,.2f}, soma total dos investimentos R$ {soma_geral['Soma Investimentos'].sum():,.2f} \nE com uma média de ROI de {soma_geral['Media ROI(%)'].mean():.2f} % ")

Porcentagem_leads=[]
soma_todos_leads = soma_geral['Soma Leads'].sum()
carro = list(df_base_formatada['Carros_Nameplate'].unique())
for c in carro:
    soma_unica_leads = df_base_formatada.loc[df_base_formatada['Carros_Nameplate']==c]['Leads'].sum()
    porcentagem_leads = (soma_unica_leads*100) / soma_todos_leads
    print('Carros | soma leads | porcentagem (%)')
    print(c, soma_unica_leads, porcentagem_leads)
    registro = [c, soma_unica_leads, porcentagem_leads]
    Porcentagem_leads.append(registro)
 

df_Porcentagem_leads = pd.DataFrame(Porcentagem_leads, columns=['Carros', 'soma leads', 'Porcentagens (%)'] )
df_Porcentagem_leads= df_Porcentagem_leads.sort_values(by='Porcentagens (%)', ascending=False)
print(f"soma da Porcentagem: {df_Porcentagem_leads['Porcentagens (%)'].sum():.2f} %")
df_Porcentagem_leads
```

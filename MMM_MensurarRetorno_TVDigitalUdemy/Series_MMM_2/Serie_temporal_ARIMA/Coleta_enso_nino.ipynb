```python
#!pip install selenium
#criar um navegador
#pesquiso no google => selenium with python
#https://selenium-python.readthedocs.io/getting-started.html
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
# criar o navegador
navegar = webdriver.Chrome()

navegar.get('https://origin.cpc.ncep.noaa.gov/products/analysis_monitoring/ensostuff/ONI_v5.php')


#conda activate Maq2


```

```python
element = navegar.find_element("xpath","/html/body/table[4]/tbody/tr/td[2]/table[2]")


html_content = element.get_attribute('outerHTML')
```

```python
print(html_content)
```

```python
import pandas as pd
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_content, 'html.parser')
#table = soup.find(name='table')
table = soup.find_all('table')
df = pd.read_html(str(table))
```

```python
df
```

```python
display(df[0])
```

```python
Basedados = pd.DataFrame(df[0])
```

```python
Basedados
```

```python
Basedados.columns = Basedados.iloc[0]
Basedados
```

```python
Basedados=  Basedados.drop(0).reset_index(drop=True) # drop linha (o)
```

```python
Basedados
```

```python
Basedados.info()
```

```python

print(list(Basedados.columns[1:]))

for numeric_columns in list(Basedados.columns[1:]):
    Basedados[numeric_columns] = pd.to_numeric(Basedados[numeric_columns], errors='coerce')
    # After replacing non-numeric values with NaN, convert to float
    Basedados[numeric_columns] = Basedados[numeric_columns].astype('float')

```

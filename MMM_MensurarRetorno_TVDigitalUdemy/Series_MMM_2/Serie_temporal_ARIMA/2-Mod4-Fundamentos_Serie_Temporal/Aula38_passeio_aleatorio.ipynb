# **CRIAÇÃO DE UM PASSEIO ALEATÓRIO**

```python
# importar nossas bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# configurar o tamanho da figura
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
# importar a biblioteca para criar amostras num determinado intervalo de valores
# criar o valor aleatório => random 
# sample amostra
from random import sample, random
```

```python
# criando uma variável dados que recebe a quantidade de amostras, de range total de 100, 
#e tiraremos 41 valores, de forma aleatória, sample/random. 
# observação, mudará sempre que rodarem
dados1 = sample(range(100), k=41)
dados1
```

```python
# criar a variável anos
# vai de 1980 até 2020, na realidade
anos = list(range(1980, 2021))
display(anos)
print(len(anos))
```

```python
# junção de dois dados array
serie1 = pd.Series(dados1, index = anos)
serie1
```

```python
serie1.plot()
plt.show()
```

```python

# Criando a figura e o gráfico
plt.figure(figsize=(10, 5))

# Plotando a série
plt.plot(serie1, label="Série 1")

# Adicionando um título ao gráfico
plt.title("Meu Gráfico de Série 1")

# Definindo rótulos dos eixos
plt.xlabel("anos")
plt.ylabel("Série")

# Rotacionando os rótulos das datas no eixo x para melhorar a visualização
plt.xticks(rotation=0)

# Adicionando a legenda
plt.legend()

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

# Abaixo está a análise do passeio aleatório agora de uma forma mais formal. Usando na criação o comando for.

```python
## Série Passeio aleatório
serie2 = list() # criar lista
serie2.append(-1 if random() < 0.5 else 1) # incluir os dados aleatórios criados
# incluir -1 se o valor gerado for menor que 0,5  caso contrarío será 1 
print(serie2)
for i in range(1, 5): # iremos fazer o processo acima por 5 vezes, somado com o dado anterior.
    movimento = -1 if random() < 0.5 else 1
    valor = serie2[i-1] + movimento
    # é que no nosso caso o nosso i está começando de 1, por conta do range 1 à 1000, por isso
    # subtraímos o valor serie2[i-1]
    serie2.append(valor) # só subtrair 
    print(valor)
print(serie2)
```

```python
## Série Passeio aleatório
serie2 = list() # criar lista
serie2.append(-1 if random() < 0.5 else 1) # incluir os dados aleatórios criados
# incluir -1 se o valor gerado for menor que 0,5  caso contrarío será 1 
for i in range(1, 1000): # iremos fazer o processo acima por 1000 vezes, somado com o dado anterior.
    movimento = -1 if random() < 0.5 else 1
    valor = serie2[i-1] + movimento
    # é que no nosso caso o nosso i está começando de 1, por conta do range 1 à 1000, por isso
    # subtraímos o valor serie2[i-1]
    serie2.append(valor) # só subtrair 
```

```python
serie2 = pd.Series(serie2)
serie2.plot()
plt.show()
```

```python

# Criando a figura e o gráfico
plt.figure(figsize=(10, 5))

# Plotando a série
plt.plot(serie2, label="Série 2")

# Adicionando um título ao gráfico
plt.title("Meu Gráfico de Série 2 é o Passeio Aleatório")

# Definindo rótulos dos eixos
plt.xlabel("anos")
plt.ylabel("Série passeio aleatório")

# Rotacionando os rótulos das datas no eixo x para melhorar a visualização
plt.xticks(rotation=0)

# Adicionando a legenda
plt.legend()

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

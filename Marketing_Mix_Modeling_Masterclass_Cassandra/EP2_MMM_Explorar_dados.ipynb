```python
# Calcula a correlação entre Sales e TV
correlacao_sales_tv = df['sales'].corr(df['TV'])
print("Correlação entre Sales e TV:", correlacao_sales_tv)

correlacao_sales_Radio = df['sales'].corr(df['Radio'])
print("Correlação entre Sales e Radio:", correlacao_sales_Radio)

correlacao_sales_FB = df['sales'].corr(df['FB'])
print("Correlação entre Sales e Facebook:", correlacao_sales_FB)
```

A melhor correlação foi a do Facebook, com o maior valor, o mais próximo de 1.

```python
import seaborn as sns

# Criar um gráfico de dispersão com linha de regressão
plt.figure(figsize=(10, 6))
sns.regplot(x='TV', y='sales', data=df, marker='o', color='b')

# Adicionar títulos e rótulos
plt.title('Relação entre Vendas e Publicidade na TV')
plt.xlabel('Publicidade na TV (custo)')
plt.ylabel('Vendas (valor)')
plt.text(10, 23, f'Correlação: {correlacao_sales_tv:.2f}', fontsize=12, color='black')

# Exibir o gráfico
plt.grid()
plt.show()
```

série de potência => 

```python
# Criar um gráfico de dispersão com linha de regressão
plt.figure(figsize=(10, 6))
sns.regplot(x='TV', y='sales', data=df, marker='o', color='b', label='Linha de Tendência')

# Adicionar a linha geométrica (y = x) que representa correlação perfeita
max_value = max(df['sales'].max(), df['TV'].max())
plt.plot([0, max_value], [0, max_value], color='r', linestyle='--', label='Linha Geométrica (y=x)')

# Adicionar títulos e rótulos
plt.title('Relação entre Vendas e Publicidade na TV')
plt.xlabel('Publicidade na TV (custo)')
plt.ylabel('Vendas (valor)')

# Exibir a legenda
plt.legend()

# Exibir o gráfico
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

# Definir variáveis independentes (X) e dependentes (y)
X = df['TV'].values.reshape(-1, 1)  # reshape para ser uma matriz 2D
y = df['sales'].values

# Criar uma transformação de polinômios de grau 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Ajustar o modelo de regressão linear
model = LinearRegression()
model.fit(X_poly, y)

# Criar um intervalo de valores para a linha de ajuste
X_fit = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_fit_poly = poly.transform(X_fit)
y_fit = model.predict(X_fit_poly)

# Plotar os dados e a linha de regressão polinomial
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Dados Reais')
plt.plot(X_fit, y_fit, color='red', label='Linha de Tendência (Série de Potência)')
plt.title('Relação entre Vendas e Publicidade na TV com Regressão Polinomial')
plt.xlabel('Publicidade na TV (custo)')
plt.ylabel('Vendas (valor)')
plt.text(10, 23, f'Correlação R²: {correlacao_sales_tv:.2f}', fontsize=12, color='black')

plt.legend()
plt.grid()
plt.show()
```

# Métricas de Desempenho*

## **Criando as métricas de desempenho do prophet**



Vamos importar também a função Performance metrics do pacote diagnostic e a função plot cross validation metric do pacote prophet

```python

from prophet.diagnostics import performance_metrics

# **Criando as métricas de desempenho do prophet**"""

import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from prophet.plot import add_changepoints_to_plot
from prophet.diagnostics import cross_validation
from prophet.diagnostics import performance_metrics
from prophet.plot import plot_cross_validation_metric
```

```python
df = pd.read_csv('online_retail.csv')
display(df)
df.columns = ['ds', 'y']
display(df)
```

estamos instanciando o nosso modelo com base nos dados online retail. 

```python

model = Prophet(yearly_seasonality=4)
model.fit(df)
forecast = model.predict()
```

```python
forecast
```

```python
fig = model.plot(forecast)
add_changepoints_to_plot(fig.gca(), model, forecast)
plt.show()
```

## Vamos executar aqui também  nosso comando de validação cruzada. 

```python

df_cv = cross_validation(model,
                        horizon='90 days',
                        period='30 days',
                        initial='730 days')
```

#### para utilizarmos a função performance métrica, nós precisamos passar como input o dataframe que foi a saída da função do Cross Validation. Pois, ele conterá os valores preditos e os valores reais para que a gente possa gerar as métricas de acerto. Então, basta passar ela par essa função  vamos ver o resultado dela. 

```python
df_p = performance_metrics(df_cv)
df_p.head()
```

Agora vamos a interpretação desse dataframe. Aqui nós vemos que temos o resultado de cada uma das métricas que nós vimos anteriormente. Repare que esse dataframe é agrupado pelo número de dias do horizonte de previsão. 

Você pode estar se perguntando de porque que o arquivo começa com 9 dias ao invés do primeiro dia do horizonte de previsão. Isso acontece porque por padrão, essa média utiliza uma janela móvel de 10% do horizonte. Como nós definimos nosso horizonte como 90 dias, os 10% seria nove (9) dias. Então, para ele gerar essa média aqui, ele precisou ter como insumo os nove(9) dias anteriores. 

Por isso que, como as linhas anteriores, não tem nove (9) dias de histórico, elas não são trazidas aqui com as métricas de avaliação. 

Nós vamos ver que isso pode ser alterado como o parâmetro rolling window. Mas, por hora, é isso que você precisa entender. Agora vou demonstrar como podemos visualizar as métricas de forma visual através da função plot_cross_validation_metric. 

#  metric='mae' métrica mae
se não especificar a métrica,  o modelo trará todos 

Aqui no eixo horizontal nós temos o horizonte de dias e no eixo vertical nós temos a escala do erro médio. Ou seja, aqui no ponto 20, por exemplo, nós temos o erro médio absoluto de todas as previões para o horizonte de 20 dias à frente do último dado da nossa série temporal. Ou seja, 20 dias no futuro. Essa linha azul que está percorrendo o gráfico, seria efetivamente a média desses pontos. E um detalhe interessante, uma outra função da média móvel é um detalhe interessante. Um dos motivos para a gente ter uma média móvel aqui, quando estamos computand a média das métricas de desempenho é justamente para suavizar essa linha azul, pois cap contrário, ela seria um pouco mais irregular e talvez dificultasse um pouco a interpretação. 

```python
fig = plot_cross_validation_metric(df_cv, metric='mae')
plt.show()
```

```python
fig = plot_cross_validation_metric(df_cv, metric='mape')
plt.show()
```

## Vamos simular abaixo, reduzindo a nossa janela móvel pela qual nós estamos fazendo a média dos pontos de erro. 

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
plot_cross_validation_metric(df_cv,
  metric='rmse',
  rolling_window=.01, #utilizando o novo método, rolling_window
  ax=ax)
plot_cross_validation_metric(df_cv,
metric='rmse',
rolling_window=.1,
ax=ax)
plt.show()
```




Nós vemos que efetivamente a média que tinha uma suavização que iniciava apenas no 9 dia do horizonte é muito mais suave do que a média que se inicia no segundo dia do horizonte. 

Então, as janelas móveis nos ajudam a deixar um pouco mais interpretável essa nossa média. 

É dessa forma que nós geramos as métricas de desempenho no prophet e que nós conseguimos visualizar também essas métricas de forma visual.

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
plot_cross_validation_metric(df_cv,
  metric='rmse',
  rolling_window=.1, #utilizando o novo método, rolling_window -> O ideal
  ax=ax)
#https://facebook.github.io/prophet/docs/diagnostics.html
plt.show()
```

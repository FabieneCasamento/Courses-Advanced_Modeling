
As séries temporais no mundo real possuem diversas mudanças em sua trajetória, de forma que puramente linear. Não é possível traçar uma reta que represente toda a natureza das séries temporais. Você deve ter notado que, em alguns exemplos, a tendência não era uma reta perfeita. Isso porque o Porphet tem a capacidade de receber mudanças no comportamento da tendência e traçar uma nova tendência a partir daquele ponto. Essas mudanças de tendências no Prophet são chamadas de change points. 


Entender como funciona o comportamento padrão do Prophet referente a essas mudanças e como podemos ajustá-las manualmente, quando entendemos que não está se comportando de forma mais adequada. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from prophet.plot import add_changepoints_to_plot
df = pd.read_csv('../Modulo5/divvy_daily.csv')
display(df)
```

```python
df = df[['date', 'rides']]
df['date'] = pd.to_datetime(df['date'])
df.columns = ['ds', 'y']
display(df)
```

#### Função auxiliar que vai nos ajudar a visualizar onde que os Change Points estão localizados: 

```python

import numpy as np

def set_changepoints(df, n_changepoints, changepoint_range):

    df = df.sort_values('ds').reset_index(drop=True)

    hist_size = int(np.floor(df.shape[0] * changepoint_range))
    if n_changepoints + 1 > hist_size:
        n_changepoints = hist_size - 1
        print('n_changepoints greater than number of observations. Using {}.'.format(n_changepoints))
    if n_changepoints > 0:
        cp_indexes = np.linspace(0,
                                 hist_size - 1,
                                 n_changepoints + 1).round().astype(int)
        changepoints = df.iloc[cp_indexes]['ds'].tail(-1)
    else:
        # set empty changepoints
        changepoints = pd.Series(pd.to_datetime([]), name='ds')
    return changepoints
```



Vamos visualizar como que o Prophet distribui o Change points em uma série temporal por padrão


Por padrão o Prophet colocará 25 pontos de mudanças potenciais espaçadas uniformemente nos primeiros 80% dos dados. 

Antes mesmo de determinar se aquele é efetivamente um ponto de mudança, repare, é um ponto de mudança potencial. Os pontos potenciais estão indicados por linhas tracejadas na vertical. 

```python
changepoints = set_changepoints(df, 25, .8)
#Por padrão o Prophet colocará 25 pontos de mudanças potenciais espaçadas uniformemente nos primeiros 80% dos dados.
changepoints
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(df['ds'], df['y'], c='#0072B2')
for cp in set_changepoints(df,25, .8):
    plt.axvline(x=cp, c='gray', ls='--')
plt.xlabel('Date')
plt.ylabel('Rides per day')
plt.show()
```

Vamos instanciar o nosso modelo com uma configuração padrão. É nessa etapa que o Prophet determinará a magnitude de mudanças de cada um desses potenciais Change Points, e irá determinar se eles efetivamente representam uma alteração repentina na inclinação da nossa tendência. 

```python
model = Prophet(seasonality_mode='multiplicative',yearly_seasonality=4)
model.fit(df)
forecast = model.predict()
forecast
```


Vamos plotar os change points oficiais.


```python

fig = model.plot(forecast)
add_changepoints_to_plot(fig.gca(), model, forecast)
plt.show()
```

### As linhas tracejadas verticais são os change points que o Prophet entendeu como significativo. 

A figura descreve que entre uma linha e outra tracejada houve uma mudança significativa na tendência. Por mais que visualmente ela não seja tão evidente matematicamente, a diferença que nós temos de inclinação entre essa reta e essa se tornou significativamente para o Porphet. 
A magnitude de cada um daqueles 25 pontos potenciais de mudança é armazenada no model params dentro do item <b>delta</b>

```python

print(model.params['delta'])
```

## Repare que esses valores foram normalizados de mode que seus valores absolutos não têm um efeito sentidos para nós, mas suas magnitudes relativas sim. 

Como esses números são calculados por meio de um processo de otimização em vez de uma equação determinística. Você pode chegar a valores exatos diferentes. Mas os expoentes sempre deve ser aproximadamente os mesmos. A maioria dessas maginitudes tem um expoente de -8 com -9, o que significa que na notação padrão você deve deslocar o ponto decimal desse número 8 ou 9 dígitos para esquerda, o que significa que esses números estão muito próximos de 0. 
Você pode visualizar essas magnitudes através desse código abaixo. 

Na imagem abaixo temos alguns elementos. 

```python

ylims = fig.gca().get_ylim()

fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'k'
add_changepoints_to_plot(ax1, model, forecast, trend=True)
ax1.set_xlabel('ds')
ax1.set_ylabel('y', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(ylims[0] + 5000, ylims[1] - 7000)
ax1.grid(False)

ax2 = ax1.twinx()

color = '#0072B2'
ax2.set_ylabel('Trend rate change')
ax2.bar(changepoints, model.params['delta'][0], width=30, color=color, alpha=.75, label='Trend rate change')
ax2.tick_params(axis='y', labelcolor='k')
ax2.axhline(y=0, c=color, ls='--')
ax2.axhline(y=.01, c='gray', ls=':')
ax2.axhline(y=-.01, c='gray', ls=':')
ax2.set_ylim(top=-ax2.get_ylim()[0], bottom=ax2.get_ylim()[0])

ax2.legend()

fig.tight_layout()
ax1.set_zorder(1)
ax1.patch.set_visible(False)
plt.show()
```



A linha sólida é a tendência; as linhas tracejadas são os chang points significativos; as barras  são os valores das magnitudes do ponto de mudança, ou seja, quanto que a inclinação de um segmento desses 25 cortes; as barras verticais são as magnitudes dos pontos de mudanças, ou seja, o quanto que aquela tendência se inclinou de um segmento para outro. 

Novamente, a maior parte dessas magnitudes é quase 0, portanto, não aparece no gráfico onde a linha tracejada horizontal azula indica um ponto 0 dessa magnitude. 



As barras que se posicionam para cima são pontos de mudança com magnitude positiva, ou seja, uma inclinação positiva da nossa tendência. E as barras que se estendem para baixo significam que a nossa tendência teve um declínio. 

Em resumo, essa é a forma que o Prophet identifica mudanças de tendência e a forma como ele trabalha por padrão.

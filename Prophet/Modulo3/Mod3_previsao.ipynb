```python
#importar a biblioteca

from prophet import Prophet
```

```python

import pandas as pd
import matplotlib.pyplot as plt
```

# carregar os dados para fazer a previsão correta. 

```python
df = pd.read_csv('co2-ppm-daily_csv.csv')
display(df)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

Teremos que renomear as colunas


```python


df.columns = ['ds', 'y']
```

```python
df
```




trainamento com o dataframe com o dado temporal

```python
model = Prophet()
model.fit(df)
```

```python
future = model.make_future_dataframe(periods = 365*10)
```

```python
future
```

Esse dataframe é a soma do período da série temporal que nós já temos, começa em 1958, e acrescenta o valor de mais 10 anos.

A previsão vai ser o predict do modelo 

```python
forecast = model. predict(future)
forecast
```

usar o método plot que tem dentro do próprio prophet, que utiliza o matplotlib 

```python
fig = model.plot(forecast, xlabel = 'Date', ylabel='CO2 PPM')
plt.show()


```

Os pontos pretos são os dados reais da série temporal,  o azul é a previsão. 

O valor azul mais claro representa o intervalo de confiança da previsão. Com os limites inferiores e superiores.

ds = timestamp, carimbo da data da nossa série temporal, no caso com dados desde 1958.

trend = tendência para aquele registo e amazena nesse campo· 

<b>yhat é uma notação estatística para designar uma variável que é predita</b> a partir de uma variável real.

ou seja é um $ \hat{y} $, que predito


Abaixo estão <b>os intervalos de confiança</b> que o prophet atribuiu para essa previsão que ele fez. 

yhat_lower $ \hat{y}_{lower}$

yhat_upper $ \hat{y}_{upper}$




O intervalo de confiança determina o nível de certeza que ele tem com relação aquela previsão que ele está dando. Conforme o tempo vai pasando ele vai aumentando aqui essa faixa do intervalo de confiança, ou seja, ele tem de certo mode menos certeza de que aquela previsão vai ser exatamente na linha azul. 

O Prophet trabalha com intervalo de confiança de 95%, ele está dizendo com isso daqui que esse valor predito tem 95% de chances de estar dentro desse intervalo aqui da área azul que a gente está visualizando.



Nós também temos um intervalo de confiança da tendência, que é o trend_lower e o trend_upper.

<b>O additive_terms, o additive_terms_lower e o additive_terms_upper são a soma do valor das sazonalidades que o Porphet está trabalhando para aquele determinado modelo</b>. Neste caso, temos a sazonalidade semanal e anual, assim <b>o additive é a soma dos dois tipos de sazonalidades, ou seja o additive_terms é o combinado delas</b>. 
Além disso, o additive está sendo usado como um parâmetros para fazer a previsão. 

O weekly e yearly são as sazonalidades, com os seus intervalos de confiança, indicados com _lower e _upper. 

Multiplicative_terms, no caso está com o valor zerado. Isso acontece por que rodamos como um modelo na modalidade aditiva da sazonalidade. O multiplicativo seria a multiplicação da tendência com relação a sazonalidade.  

yhat, que é o valor predito para um determinado timestamp

### Agora vamos entender como nós podemos visualizar e interpretar as informações que o Prophet está utilizando para projetar a nossa série temporal. Vamos fazer isso através de uma função que a biblioteca Prophet contém. 

```python
fig2 = model.plot_components(forecast)
fig2.show()
```

na primeira figura do arquivo, mostra a série decomposta, a prevista ais a real, isolando a tendência e isolando as sazonalidades que constituem essa série temporal.

Nesta segunda figura apresentamos a tendência, a sazonalidade semana e anual. 

Verificamos se os dados tem uma tendência de crescimento ou de baixa. 

No caso, a sazonalidade semanal no gráfico apresenta valores muito baixos.Isso não necessariamnte está sendo significante para o Prophet fazer a previsão. Isso aqui pode se retratar apenas de algum ruído que o dado possui e então a gente vai ver mais para frente como o modelo pode desconsiderar essa sazonalidade.  

Como o nosso dado é de nível diário,  o Prophet acaba procurando inferir uma sazonalidade semana, mas nem sempre ela existe. Como parece o caso. 



Já na sazonalidade anual, tem uma variação significante definida ao longo do ano ou durante o ano. Então essa sazonalidade está sendo com certeza importante para o Prophet fazer a previsão da série temporal, e ela deve ser mantida.

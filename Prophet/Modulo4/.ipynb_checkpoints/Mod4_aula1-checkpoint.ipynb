Por padrão, o Prophet trabalha com os dados diários e inicialmente a ferramenta foi projetada para trabalhar com dados diários. Mas, conforme ela foi crescendo, outras granularidades foram sendo incrementadas para que ela pudesse projetar. 

Há algumas peculiaridades que a gente tem que levar em consideração, que será abordado.


Teremos como base os dados do Air Passenger. 


```python
import pandas as pd
df_air_passengers = pd.read_csv('AirPassengers.csv')
display(df_air_passengers)
```

primeira etapa é transformar a coluna de mês en data (timestamp) especificamente. 

E a segunda transformação é modificar o nome das colunas. 

O prophet precisa que a coluna de data esteja como ds e a coluna da série temporal esteja como y.

```python
df_air_passengers['Month'] = pd.to_datetime(df_air_passengers['Month'])
df_air_passengers.columns =['ds','y']
df_air_passengers
```

```python
df_air_passengers.info()
```

```python
#importar a biblioteca

from prophet import Prophet
#instânciado o modelo
model = Prophet(seasonality_mode ='multiplicative')
```

```python
model.fit(df_air_passengers)
```

```python
# prever os próximos 5 anos
future = model.make_future_dataframe(periods = 365 * 5)
```

```python
future
```

### colocar as datas na função predict para prever

```python



forecast = model.predict(future)
forecast
```

# Visualizar o Forecast

```python
# Visualizar o Forecast
import matplotlib.pyplot as plt
fig = model.plot(forecast)
plt.show() 
```

Os dados se ajustaram bem enquanto se tinha o dado de treinamento. Mas, durante a previsão, durante a parte que não tem dado de treinamento, ele não conseguiu se ajustar bem. Nós vemos que visivelmente não está coerente essa previsão que o modelo está fazendo. 

### Isso tudo se deve justamente ao dado não estar na escala mensal. Então, como ajustamos isso?

O ajuste deverá ser feito no make_future_dataframe(), nos devemos incluir o parâmetro freq, com a informação de 'MS', que são dados mensais, e significa Month Start. Esse parâmetro vai receber qualer coisa que o pandas entenda como uma frequência válida. 

### refazendo com o novo parâmetro

```python
# e respeita o que o pandas entende como frequências válidas. 
future = model.make_future_dataframe(periods = 365 * 5, freq= 'MS')
future
```

```python
future.info()
```

```python
#outro ajuste é nos períodos , em vez do período ser em dia 365 * 5, colocamos como dados mensais, 12*5

future = model.make_future_dataframe(periods = 12 * 5, freq= 'MS')
future
```

```python
forecast = model.predict(future)
forecast
```

# O prophet agora ajustou muito bem os dados 

```python
# Visualizar o Forecast
import matplotlib.pyplot as plt
fig = model.plot(forecast)
plt.show() 
```

A previsão está muito coerente com os dados históricos, que nós tínhamos nessa seérie temporal. 

Ajustes que nós vamos trabalhar com uma frequência diferente da diária. Precisamos aletar a quantidade de entadas do dataframe futuro que nós vamos passar. 

Basicamente, a gente precisa alterar a quantidade de entradas do dataframe futuro que nós vamos passar para a função predict. Nós conseguimos controlar no make_future_dataframe. Antes havia dado errado, pois havimaos passado para o Prophet prevê períodos que el não tinha treinamento para fazer a previsão corretamente. O modelo não dá erro, ele gera resultado incoerente. 

Assim, quando nnós trabalhamos com frequências diferentes da diária, precisamos alterar a função make_future_dataframe.

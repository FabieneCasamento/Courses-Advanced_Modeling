# Adicionando os Regressores Adicionais



Fundamentalmente, grande parte do trabalho de previsão de séries temporais é feito utilizando informações da própira série temporal, mas podemos enriqueCê-la com informações extras também. Já tivemos contato com regressões adicionais de forma discreta através da adição de feriados, pois os feriados no Prophet são implementados como um regressor binário. A primeira coisa a considerar como regressores adicionais, sejam binários ou contínuos, é que você deve ter valores futuros conhecidos para todo o período de previsão. Isso não é um problema com feriados, porque sabemos exatamento quando ocorrerá a cada feriado no futuro. Todos os valores futuros devem ser conhecidos. Como acontece com os feriados ou devem ter sidos previstos separadamente. 


Porém, você deve ter cuidado ao criar uma previsão usando dados que já foram previstos, pois o erro da primeira previsão aumentará o erro na segunda previsão e os erros se acumularão continuamente. 

Entretanto, se uma variável for muito mais fácil de prever que outra, então este pode ser um caso em que essas variáveis empilhadas, quando previstas, fazem sentido. Uma série temporal hierárquica é um exemplo de caso em que isso pode ser útil. Você pode encontrar bons resultados prevendo os valores diários mais confiáveis do que uma série temporal, por exemplo, usando as os valores para prever horários de uma outra série temporal que seja mais difícil de prever. 

E para aprender esse tópico, nós continuaremos utilizando a base divvy daily bikes, que possui duas colunas extras, que até então não estávamos utilizando.  

```python

import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
df = pd.read_csv('../Modulo5/divvy_daily.csv')
df
```

Vamos fazer uma exploração rápida, utilizando o campo que nós vamos adicionar como regressor, que no caso vai ser o campo weather;

```python
print(df.groupby('weather')['weather'].count())
```

### Esse comando está contando quantos registros a gente tem para cada valor distinto desse campo weather. 


Esse informação do campo wheather é do tipo categórico e por isso representá-la em nossos dados de forma binária. Ou seja, cada valor contido em moeda se torna uma coluna que vai receber 1 quando aquele dia tiver aquela classificação e 0 quando não tiver. Nós conseguimos fazer isso de forma bem simples, no pandas, através do comando get_dummies, onde passamos a coluna e automaticamente é criada as colunas dummies para nós. 

```python
df['date'] = pd.to_datetime(df['date'])
#vamos renomear as colunas
df.columns = ['ds', 'y', 'temp', 'weather']
df
```

```python
df = pd.get_dummies(df, columns=['weather'], prefix='',
                    prefix_sep='')
```

A exemplo de leitura do nosso dataframe para esse dia, o nosso tempo estava com chuva ou com neve. Perceba que só no primeiro dia está com a coluna True, neste exemplo abaixo. 

```python
df
```

```python
df.head()
```

Agora, vamos instanciar o nosso modelo, definindo o modo de sazonalidade multiplicativa e a sazonalidade anual, com uma ordem de 4, como fizemos nos capítulos anteriores.

```python

model = Prophet(seasonality_mode='multiplicative',
yearly_seasonality=4)
```

Antes de instanciar, nós precisamos adicionar cada um dos regressores separadamente através dessa chamada de add_regressores, assim passamos parâmetros personalizados apenas para esse primeiro regressor, e os utros estão assumndo os valores padrões. 

```python
model.add_regressor(name='clear',
prior_scale=10,
standardize='auto',
mode='multiplicative')

model.add_regressor('not clear')

model.add_regressor('rain or snow')
```

Não que isso seja necessário para essa série, mas apenas para exemplificar as possibilidades. E nisso há um ponto interessante. Observe que nós estamos deixando o regressor 'cloudy' para fora e estamos fazendo isso intencionalmente para que o Prophet obtenha resultados mais precisos na previsão. Isso não é estritamente necessário, no entanto, como a inclusão de todas as 4 variáveis binárias, a gente poderia introduzir um efeito chamado multicolinearidade. Isso faz com que seja difícil interpretar o efeito individual de cada condição e para isso excluímos uma delas.

No entanto, o Prophet é bem robusto a multicoliearidade em regressões adicionas, portanto, não deve afetar significativamete os resultados finais. E para entender um pouquinho sobre essa multicolineridade, quando todas as colunas estão presentes, pode haver uma alta correlação entre elas, o que pode dificultar a interpretação dos efeitos individuais de cada condição. Com isso, é uma prática comum dentro de regressões, quando estamos modelando uma variável categórica, nós deixamos uma condição de fora.


Uma opção seria deixar o própiro get_dummies, que executamos anteriormente, definir isso para nós. Se antes, a gente especificasse um parâmetro drop first, ele ia excluir uma das condições automaticamente. Mas, decidimos não fazê-lo para que pudéssemos escolher qual coluna excluir. Pois, essa condição nublada (cloudy) é de longe a mais frequente. Portanto, ao excluí-la estamos essencialmente afirmando que a cloudy é a condição do tempo padrão, e todas as demais serão declaradas como os desvios.  

```python

from datetime import timedelta
# Remove final 2 weeks of training data
train = df[df['ds'] < df['ds'].max() - timedelta(weeks=2)]
train
```


Lembrando que agora precisamos de dados futuros, para os regressoes adicionais. Para simplificar o nosso exemplo, vamos tirar simplesmente o período de dados de treinamento já existente. No caso, vamos tirar duas semanas. Então, nesse bloco, estamos simplesmente filtrando o dataframe, removendo as últimas uas semanas dele. Com isso, nossos dados de treinamento ficam duas semanas curtos e vamos utilizar a informação do tempo no nosso dataframe futuro também.

```python
model.fit(train)
future = model.make_future_dataframe(periods=14)
future
```

```python
future['clear'] = df['clear']
future['not clear'] = df['not clear']
future['rain or snow'] = df['rain or snow']
future
```

```python

forecast = model.predict(future)
fig2 = model.plot_components(forecast)
plt.show()
```

Com isso, uma nova informação apareceu no nosso gráfico de componentes, que é o impacto das variáveis que é o impacto das variáveis que adicionamos na previsão. O valor 0 de referência se torna, no nosso caos, o dia nublado, que á condição padrão e todos os outros efeitos são decorrentes de alguma das outras opções. Por exemplo, no início, como porcentagem positiva, são dias com o aumento, são dias que estavam com o tempo clear, ou tempo mais limpo, e os que têm uma redução na utilização, são os dias em que estão com o tempo 'Rain' ou 'Snow'. Assim, dessa forma conseguimos mensurar o impacto dessas categorias no decorrer da série temporal, dessa forma que adicionamos regressores categóricos em nosso modelo do Prophet através da transformação em informações binárias. 

#  Regressores Continuos

## Vamos aprender como adicionar regressões de variáveis contínuas. 

Por exemplo, vamos pegar o modelo anterior, e vamos adicionar a informação de temperatura. 

# **Adicionando Regressores Continuos

O processo de adicionar variáveis contínuas no modelo é o mesmo da variável categórica. Fazemos isso através do método add_regressor, mas nós não precisamos nos preocupar como gerar as variáveis dummies, o que torna o processo mais simples. 

Adicionado a temperatura no add_regressor, assim como na previsão futura. 

```python
model = Prophet(seasonality_mode='multiplicative',
                yearly_seasonality=4)
model.add_regressor('temp')
model.add_regressor('clear')
model.add_regressor('not clear')
model.add_regressor('rain or snow')

model.fit(train)
```

```python
model
```

```python
df.tail(10)
```

```python
future = model.make_future_dataframe(periods=14)
future['temp'] = df['temp']
future['clear'] = df['clear']
future['not clear'] = df['not clear']
future['rain or snow'] = df['rain or snow']
future
```

```python
forecast = model.predict(future)
```

```python
forecast
```

```python
fig2 = model.plot_components(forecast)
plt.show()
```

Executando esse modelo, nós podemos ver que essa regressor multiplicativa possui uma riqueza de informações muito maior. Nós vemos que a sazonalidade da variação da temperatura no dcorrer das estações do ano estão representadas, na última figura, servindo de insumo para o modelo. Nós podemos ver que os nossos dados de regressões acabam se adicionando, colaborando aqui com a nossa sazonalidde, adicionando mais impacto na variação da utilização. 
Então nós vemos que a sazonalidade anual chega em picos de 30% e de menos -40% e menos de -30%. (terceira figura). 

Nós olhamos agora pela ótica da temperatura, última figura, isso tem uma amplitude maior. Então é dessa forma que adicionamos regressores contínuos no nosso modelo Prophet.
